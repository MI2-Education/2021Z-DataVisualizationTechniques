```python
import numpy as np
import matplotlib.pyplot as plt

lata = np.arange(2003, 2021, 1).tolist()
golenamecz = [0.00,0.44,0.18,0.43,0.50,0.12,0.14,0.27,0.88,0.38,1.11,0.56,0.60,1,1,0.86,1.4,0.67]
for i in range(len(lata)):
    lata[i]=str(lata[i])

srednia = 0
for i in range(len(lata)):
    srednia+=golenamecz[i]
srednia=srednia/len(lata)

fig = plt.figure(figsize=(12, 6))
fig2 = fig.add_axes([0,0,1,1])
fig2.bar(lata, height = golenamecz, width = 0.6, color = "black")
fig2.hlines(srednia, 0, len(lata)-0.25, colors = "r")
plt.text(0,srednia+0.01,"Średnia:",color="r")
plt.text(1,srednia+0.01,"{:.2f}".format(round(srednia, 2)),color="r")
plt.ylabel("Średnia goli na mecz")
plt.xlabel("Rok")
plt.title("Średnia goli na mecz Cristiano Ronaldo", fontsize=16)
plt.xticks(lata)
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')
for i in range(len(lata)):
    plt.text(i-0.25,golenamecz[i]+0.01,"{:.2f}".format(round(golenamecz[i], 2)))
plt.show()
```

Zastosowałem "żywsze" kolory i tym razem nie dodawałem siatki, która psuje odbiór, dodałem linię średniej, ujednoliciłem wartości zaokrąglając wszystkie do dwóch miejsc po przecinku.
