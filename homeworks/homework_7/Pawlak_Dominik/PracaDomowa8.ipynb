```python
Years = np.linspace(2007, 2020, 14)
Values = [50.0, 60.2, 64.8, 66.5, 69.2, 70.8, 74.8, 
             79.0, 79.8, 84.6, 90.4, 92.6, 97.6, 109.0]



plt.figure(figsize=(12, 8))
plt.bar(Years, Values)
plt.xlabel("Rok", fontdict = {'fontsize': 15})
plt.title("Publiczne wydatki na sluzbe zdrowia na przestrzeni lat (w mld zl)", 
          fontdict = {'fontsize': 20})
plt.ylabel("Kwota (mld zl)", fontdict = {'fontsize': 15})
plt.grid(axis = 'y')
plt.xticks([2010, 2015, 2020])
plt.ylim(0, 125)
for i in range(0, len(Years)-1):
    plt.text(Years[i] - 0.3, Values[i] + 2, Values[i], fontsize = 13)
plt.text(Years[13] - 0.4, Values[13] + 2, Values[13], fontsize = 13)
plt.show()
```
