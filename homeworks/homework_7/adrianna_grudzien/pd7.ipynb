### Wykres poprawiony w ramach pd7

Najpierw importujemy wszystkie potrzebne biblioteki, zapisujemy dane z oryginalnego wykresu i dopiero wtedy tworzymy poprawiony wykres.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
liczba_ptakow = [17850, 8595, 8601, 10503, 1295, 14524, 21138, 15043, 11380, 8860, 6721, 6453]
rok = [i for i in range(2008,2020)]
wroble = pd.DataFrame(data = {"rok" : rok, "liczba wróbli" : liczba_ptakow})
```

```python
wroble.head() #zobaczmy, jak wygląda utworzona ramka danych
```

```python
sns.set(rc={'figure.figsize':(11,6)})
sns.set_style("whitegrid") #ustawiam tło na poziome linie
```

```python
sns.barplot(data=wroble,x=rok,y=liczba_ptakow, color="green")
plt.title("Liczebność wróbla domowego na podstawie danych \nz Zimowego Ptakoliczenia w Sosnowcu w latach 2008-2019",loc="left",fontsize=15)
plt.xticks(rotation=45) #przekręcam napisy na OX
plt.xlabel("rok",fontsize=13)
plt.ylabel("liczba ptaków",fontsize=13)
plt.show()
```

### Dlaczego ostatni wykres jest lepszy?

W ostatnim wykresie zostały wprowadzone znaczące poprawki. Jest on lepszy od wykresu oryginalnego oraz tego poprawionego w pd2, ponieważ:
* zostało zlikwidowane kolorystyczne rozróżnienie słupków (które i tak niczego nie wnosiło);
* nazwy osi nie zawierają nieestetycznych podkreślników;
* tło zostało zmienione na białe z poziomymi liniami, co ułatwia odczytywanie danych (nie rzuca się tak w oczy jak na wykresie z pd2 i nie ma pionowych linii, które i tak niczego nie wnoszą);
* podpisy na osi OX są znacznie czytelniejsze (nie zlewają się w jeden, trudny do odczytania, ciąg liczb).
