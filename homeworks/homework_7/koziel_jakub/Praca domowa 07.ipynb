### Ponownie poprawiana wizualizacja

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set(rc={'figure.figsize':(16,9)})
sns.set(font_scale = 1.3)
sns.set_style("whitegrid")

etykiety_x = [
      "Under 30",
      "30-39",
      "40-49",
      "50-59",
      "60-69",
      "70-79",
      "80-89",
      "Over 90"]
zgony_sierpien = [0, 11, 26, 67, 155, 328, 523, 272]
zgony_lipiec = [0, 3, 12, 27, 58, 114, 211, 104]


df = pd.DataFrame(zip(etykiety_x,  zgony_lipiec, zgony_sierpien,), columns = ['etykiety_x' , 'July' , 'August'])
df = df.melt(id_vars='etykiety_x', var_name="Deaths in respective month")


ax = sns.barplot(data = df, x = "etykiety_x", y='value', hue="Deaths in respective month")


ax.set_title("Deaths due to COVID-19 in each group of age",fontsize=25)
ax.set_xlabel("Age", fontsize=20)
ax.set_ylabel("Number of deaths", fontsize=20)


plt.show()
```
