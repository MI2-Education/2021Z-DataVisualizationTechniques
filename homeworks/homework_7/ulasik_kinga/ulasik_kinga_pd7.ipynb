### Moja poprawka

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np

# opcje wyswietlania
sns.set(rc={'figure.figsize':(12,7)})
sns.set_style("darkgrid")
pd.options.display.max_rows=10
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)
```

```python
PercentageStud = [35.8, 2.1, 26.1, 13.4, 17, 5.2, 0.4]
Ethnicity = ["Asian", "Black", "Hispanic/Latino", "White","Non-Resident Alien", "Ethnicity Unknown", "Other"]
PercentageFac = [23.8, 3.3, 17, 44.9, 5.9, 4.5, 0.6]
```

```python
data_stud = {"Ethnicity": Ethnicity, "PercentageStud":PercentageStud}
Students = pd.DataFrame(data_stud)
data_fac = {"Ethnicity":Ethnicity, "PercentageFac":PercentageFac}
Faculty = pd.DataFrame(data_fac)

Students = Students.sort_values(by=['PercentageStud'], ascending = False) 
Faculty = Faculty.sort_values(by=['PercentageFac'], ascending = False)
```

```python
sns.barplot(data = Students, x = "Ethnicity", y = "PercentageStud", color = "mediumpurple" )
plt.title('Ethnic diversity of Undergraduate Students at University of California - Irvine')
plt.ylabel('Percentage of students [%]')
plt.show()

sns.barplot(data = Faculty, x = "Ethnicity", y = "PercentageFac", color = "mediumpurple" )
plt.title('Ethnic diversity of Faculty at University of California - Irvine')
plt.ylabel('Percentage of faculty [%]')
plt.show()
```
