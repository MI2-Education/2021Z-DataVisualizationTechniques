#### Import 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.set_option("display.notebook_repr_html", False)
```

#### Praca z danymi

```python
data = pd.read_csv("C:\\Users\\pauli\\TWD\\pd7\\covid-data.csv")
```

```python
locations = ["United Kingdom", "Germany", "Italy", "Portugal", "Poland", "Austria", 
             "Italy", "Australia", "Romania", "Czech Republic",  "Netherlands", 
             "Switzerland", "Slovakia", "Belgium", "Israel"]
df = data.loc[(data.location.isin(locations)) & (data.date == "2020-10-11"), ["location", "total_tests"]].reset_index(drop = True)
result = df.sort_values(by = ['total_tests'], ascending = 1)
result.total_tests = result.total_tests/1000000
data_for_plot = result.round(2)
```

#### Tworzenie wykresu

```python
plt.barh(data_for_plot.location, data_for_plot.total_tests)
plt.ylabel('country', size = 13)
plt.xlabel('tests in millions', size = 13)
plt.xlim(0,27)
plt.title("Overall tests in countries on November 9th", loc = "center", size = 20)


for index, value in enumerate(data_for_plot.total_tests):
    plt.text(value, index, str(value))
    
plt.show()
```
