# Wykres poprawiony v2

```python
Procent = [52, 22, 15, 5, 4,2]
pojazdy = ["Tramwaj", "Samochod", "Rower", "Autobus",
             "Poruszanie\nsie pieszo", "Hulajnoga\nelektryczna"]

plt.figure(figsize=(15,7.5))
ax = plt.axes()        
ax.yaxis.grid()#linie równoległe
ax.set_axisbelow(True)#schowanie lini pod słupkami
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.bar(pojazdy, Procent, color = "#69b3a2", edgecolor='black', width=0.71)

plt.title('Ktory środek transportu ocenia Pani/Pan jako najbardziej efektywny w Poznaniu?', fontsize=15)
plt.xticks(fontsize=13)
plt.yticks(np.arange(0, 55, 5),fontsize=13)
plt.text(-0.76,54, "%", fontsize=12)


plt.show()
```

# Podsumowanie

Wykres ten jest lepszy od wykresu oryginalnego, ponieważ został usuniety efekt 3D, ktory utrudnia odczyt wykresu, oraz nadmiar kolorów, które nie pełnią żadnej funkcji.

W porównaniu do pierwszego poprawionego został zmnienony kolor słupków, który wcześniej był bardzo intensywny i przyćmiewał opisy. Zmieniłam także proporcje słupków w stosunku do odległości między nimi(w poprzednim wykresie słupki były za szerokie, a odstępy za wąskie). Zostawiłam tylko równległe linie siatki, ponieważ w wykresie słupkowym pionowe linie nie są potrzebne. Usunełam niepotrzebne linie ramki oraz wartości wypisane nad słupkami, ponieważ można je łatwo odczytać z osi.
