# Poprawiona wizualizacja w Pythonie

```python
import pandas as pd
#import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
df = df.sort_values(by=['numer'], ascending=False)
```

```python
from matplotlib.pyplot import figure

df_always = df.loc[df['czestosc'] == 'Always']
df_often = df.loc[df['czestosc'] == 'Often']
df_some = df.loc[df['czestosc'] == 'Sometimes']
df_never = df.loc[df['czestosc'] == 'Never']

fig = plt.figure()
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1) # 2 wiersze, 1 kolumna
fig.set_figheight(12)
fig.set_figwidth(16)
fig.suptitle('How often do you work under time pressure?', fontsize = 30, y = 0.94)

ax1.bar(df_always["GEO"], df_always['Value'], color = 'darkorange')
ax1.set_ylim([0, 60])
ax1.set_ylabel('Percentage of people', Fontsize = 12)
ax1.set_title('Always', fontsize = 20, x = 0, horizontalalignment = 'left')
ax1.grid()
ax1.set_axisbelow(True)
#ax1.tick_params(rotation=90)
ax1.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)

ax2.bar(df_often["GEO"], df_often['Value'], color = 'chocolate')
ax2.set_ylim([0, 60])
ax2.set_ylabel('Percentage of people', Fontsize = 12)
ax2.set_title('Often', fontsize = 20, x = 0, horizontalalignment = 'left')
ax2.grid()
ax2.set_axisbelow(True)
#ax2.tick_params(rotation=90)
ax2.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)

ax3.bar(df_some["GEO"], df_some['Value'], color = 'deepskyblue')
ax3.set_ylim([0, 60])
ax3.set_ylabel('Percentage of people', Fontsize = 12)
ax3.set_title('Sometimes', fontsize = 20, x = 0, horizontalalignment = 'left')
ax3.grid()
ax3.set_axisbelow(True)
#ax3.tick_params(rotation=90)
ax3.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)

ax4.bar(df_never["GEO"], df_never['Value'], color = 'lightskyblue')
ax4.set_ylim([0, 60])
ax4.set_xlabel('European countries', Fontsize = 15)
ax4.set_ylabel('Percentage of people', Fontsize = 12)
ax4.set_title('Never', fontsize = 20, x = 0, horizontalalignment = 'left')
ax4.grid()
ax4.set_axisbelow(True)
#ax4.tick_params(rotation=45, horizontalalignment ="right")
plt.xticks(rotation = 45, ha="right")

#plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

## Uzasadnienie
Tak jak w wykresie ggplot2 rozdzieliłem 4 kategorie oryginalnej wizualizacji na 4 wykresy. Dzięki temu zwiększona jest czytelność. Można znacznie łatwiej odczytywać wartości słupków i porównywać je między sobą. W Pythonie postanowiłem jednak umieścić wszystkie wykresy jeden nad drugim. W ten sposób słupki odpowiadające poszczególnym krajom są w jednej pionowej osi i można je łatwo połączyć. Sprawiło to, że można było usunąć zbędne podpisy osi x. Dodatkowo, dzięki temu wizualizacja jest szersza i łatwiejsza do przeczytania.\
Oczywiście wszystkie 4 wykresy są tak samo wyskalowane i osie x mają taką samą kolejność.\
Postanowiłem zachować kolorystykę, bo mimo, że nie jest niezbędna, to podkreśla ona podział na kategorie i nawiązuje do oryginalnego wykresu.\
Moim zdaniem ta wizualizacja ma wszytkie zalety poprzedniej, zrobionej w ggplot2, ale jeszcze ją polepsza.
