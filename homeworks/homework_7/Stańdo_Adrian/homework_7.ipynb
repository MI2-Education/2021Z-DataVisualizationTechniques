## Moja nowa wizualizacja

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [12, 9]

miesiace = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August']
rok2019 = np.array([30.5, 29, 28, 31.8, 32.9, 33.7, 33, 31.3]) * 10**3
rok2020 = np.array([30.2, 27.9, 31.4, 39.2, 34.6, 31.7, 28.6, 28.2]) * 10**3

barWidth = 0.3

r1 = np.arange(len(rok2019))
r2 = [x + barWidth for x in r1]

plt.bar(r1, rok2019, width = barWidth, color = '#5a9bd4', edgecolor = 'black', capsize=7, label='2019', zorder=3)
plt.bar(r2, rok2020, width = barWidth, color = '#faa75b', edgecolor = 'black', capsize=7, label='2020', zorder=3)

plt.xticks([r + barWidth/2 for r in range(len(rok2019))], miesiace)

plt.suptitle("Total Average Monthly TV Audience '19 vs. '20", fontsize = 16, fontweight="bold", y = 0.94)
plt.title("TOTAL ADULTS (15 YEARS+)", fontsize = 10)
plt.ylabel('Number of viewers', fontweight="bold", fontsize = 12)
plt.xlabel('Month', fontweight="bold", fontsize = 12)
plt.legend()
plt.grid(zorder=0, axis="y")

plt.show()    
```

## Dlaczego nowy wykres jest lepszy od oryginału?

* Oś OY zaczyna się w 0, a nie w 20000. Widzimy odchylenie ilości widzów w kwietniu, jednak nie jest ono tak duże, jak na oryginale.
* Barwy na oryginalnym wykresie są zimne i znajdują się blisko na palecie barw - w przeciwieństwie do poprawionych wykresów.
* Tło zostało zmienione na białe - takie jakie tło strony, na której znalazł się wykres.

## Dlaczego nowy wykres w Pythonie jest lepszy niż wykres w ggplot2?

* Nie ma etykiet na każdym słupku - teraz wykres wygląda przejrzyściej.
* Wykres w Pythonie ma czarną ramkę, do której początkowo nie byłem przychylny, ale dzięki niej wykres wygląda lepiej.
* Słupki mają czarne obramowanie - tworzą kontur i uwidaczniają granice między tłem a słupkiem.
* Wartości osi OY są podane "w pełnym rozmiarze" - bez dodatkowych oznaczeń typu "K". Dzięki temu odrazu widać rząd wielkości liczb, a nie ładne małe liczby i dalej za literką ukryty mnożnik. Uważam, że trzy zera w tysiącach to nie jest zbyt dużo, żeby zapisać je w pełnej formie.
* Mamy pomocnicze linie pomagające określić wartość słupka.
* Ogólnie wydaje mi się, że nowy wykres stworzony w Pythonie wygląda dużo zgrabniej.
