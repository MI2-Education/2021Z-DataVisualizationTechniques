# Nowe wykresy

```python
tablica = {'ethnic':["Asian","Black","Hispanic/Latino","White","Non-Resident Alien","Ethnicity Unknown","Other"],
           'stud':[0.358,0.021,0.261,0.134,0.17,0.052,0.004],
           'fac': [0.238,0.033,0.17,0.449,0.059,0.045,0.006]}
ramka = pd.DataFrame(data = tablica)
plt.figure(figsize = (13,9))
sns.set(font_scale = 1.4)
sns.set_style("white")
p1 = sns.barplot(data = ramka, x = 'ethnic', y = 'stud', color = 'grey')
p1.set(xlabel='', ylabel='')
p1.set_xticklabels(p1.get_xticklabels(), rotation = -30)
ticks = p1.get_yticks()
p1.set_yticklabels(['{:,.0%}'.format(x) for x in ticks])
plt.title("Ethnic diveristy of Undergraduate students at University of California- Irvine", size = 20) 
plt.show()
```

```python
plt.figure(figsize = (13,9))
sns.set(font_scale = 1.4)
sns.set_style("white")
p1 = sns.barplot(data = ramka, x = 'ethnic', y = 'fac', color = 'grey')
p1.set(xlabel='', ylabel='')
p1.set_xticklabels(p1.get_xticklabels(), rotation = -30)
ticks = p1.get_yticks()
p1.set_yticklabels(['{:,.0%}'.format(x) for x in ticks])
plt.title("Ethnic diveristy of Faculty at University of California- Irvine", size = 20) 
plt.show()
```
