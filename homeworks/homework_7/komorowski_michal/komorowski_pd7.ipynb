### Wykres w Pythonie

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("data.csv", sep = ";")
df.head()
```

```python
fig, ax = plt.subplots(1, 1, figsize = (8, 8), dpi = 96)
sns.set(font_scale=1.4)
plt.title("Number of murders commited using firearms", x= 0.5, y = 1.05, fontsize = 16)
plt.suptitle("Gun deaths in Florida", x= 0.5, y = 1, fontsize = 26, fontweight="bold")
sns.set_style("whitegrid")
sns.lineplot(data = df, x = "Year", y = "Murder by firearm", color = "black", linewidth = 2.5, zorder = 1)
sns.scatterplot(data = df, x = "Year", y = "Murder by firearm", color = "black", s=125, zorder = 1.25)
ax.fill_between(df["Year"], df["Murder by firearm"], color="red", alpha=0.4, zorder = 0.6)
ax.set_ylabel('')    
ax.set_xlabel('')
ax.set(ylim=(0, 1000))
ax.set(xlim=(1989, 2013))
plt.hlines(721, 0, 2013, linestyle="dashed", color="gray", zorder = 0.99, alpha = 0.4)
plt.hlines(873, 0, 2013, linestyle="dashed", color="gray", zorder = 0.99, alpha = 0.4)
ax.text(1987.2, 716, "721", fontsize=14)
ax.text(1987.2, 865, "873", fontsize=14) 

plt.show()
```

### Uzasadnienie

Myślę, że ostatni wykres estetycznie prezentuje się lepiej od dwóch pozostałych. Dodatkowo zamiast tekstu nad skrajnymi punktami, dodałem przerywane linie i dodatkowe oznaczenie na poziomej osi.
