```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# było stworzone w R
usa = np.array([12170, 16344, 21454, 25257, 28856, 31662, 35790, 38249, 29181,
                23771, 23684, 25037, 30313, 35160, 44572, 45437, 54724, 58402,
                62829, 65203, 70945, 75561, 82013, 85076, 74909, 68480, 60221, 52642, 43508])

world = np.array([92788, 82560, 105042, 113759, 139210, 126390, 153401, 170997, 170943, 187693, 203045, 193016, 
                  209403, 226931, 237585, 253324, 251598, 273870, 281062, 307992, 296520, 324475, 340387, 340377, 
                  347467, 347353, 371331, 387976, 400516])

dates = np.array(["2020-03-28", "2020-04-04", "2020-04-11", "2020-04-18", "2020-04-25", "2020-05-02", "2020-05-09", 
                  "2020-05-16", "2020-05-23", "2020-05-30", "2020-06-06", "2020-06-13", "2020-06-20", "2020-06-27", 
                  "2020-07-04", "2020-07-11", "2020-07-18", "2020-07-25", "2020-08-01", "2020-08-08", "2020-08-15", 
                  "2020-08-22", "2020-08-29", "2020-09-05", "2020-09-12", "2020-09-19", "2020-09-26", "2020-10-03", 
                  "2020-10-10"])

# skalowanie względem populacji 
usa = usa/328.2
world = world/7800
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)
fig.set_figwidth(15)
fig.suptitle('7 Day Rolling Average New Coronavirus Cases Per Million')

ax1.plot(dates, usa, label='USA')
ax1.plot(dates, world, 'r', label='World')
ax1.legend()
ax1.set_xticks(np.arange(0, 29, 2))
ax1.set_xticklabels(dates[0:len(dates):2], rotation=45)
ax1.set_title('USA and World', loc='left')

ax2.plot(dates, world, 'r')
ax2.set_xticks(np.arange(0, 29, 2))
ax2.set_xticklabels(dates[0:len(dates):2], rotation=45)
ax2.set_title('World', loc='left')


plt.subplots_adjust(bottom = -3.5)



plt.show()
```

Po 1 projekcie już wiemy, że dane dotyczące populacji należy skalować względem liczności tych populacji, dlatego uznałem, że lepiej pokazać ilość zakażeń na milion osób, zamiast podawać ich dokładną wartość.
