## Końcowy wygląd

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick
```

```python
Month = ["Sep '19", "Oct '19", "Nov '19", "Dec '19", 
        "Jan '20", "Feb '20", "Mar '20", "Apr '20", 
        "May '20", "Jun '20", "Jul '20", "Aug '20", "Sep '20"]
UnemploymentRate = [3.5, 3.6, 3.5, 3.5, 3.6, 3.5, 
        4.4, 14.7, 13.3, 11.1, 10.2, 8.4, 7.9 ]
```

```python
fig = plt.figure(figsize=(15, 7))
ax = fig.add_subplot(1, 1, 1)
plt.title('Monthly unemployment rate in the US', fontsize=21, ha = 'center')

ax.bar(Month, UnemploymentRate)
rects = ax.patches
plt.ylim(0, 16)
plt.yticks(np.arange(0, 17, step = 2.5))
plt.ylabel("Unemployment rate", size = 11)

i = 0
for rect in rects:
    y_value = rect.get_height() + 0.6
    x_value = rect.get_x() + rect.get_width() / 2
    lab = str(UnemploymentRate[i]) + "%"
    plt.annotate(lab, 
                (x_value, y_value), 
                ha='center', 
                va = 'top', 
                size = 12)
    i += 1
    

    
ax.grid(color='gray', linestyle='-', 
         linewidth=0.3, 
         axis = 'y', 
        )
    
ax.set_axisbelow(True)
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
plt.xticks(rotation = 45, size=11)

plt.show()
```
