```python
import matplotlib.pyplot as plt
from matplotlib import ticker
import numpy as np
```

```python
countries = np.array(['USA', 'Wielka Brytania', 'Niemcy', 'Hiszpania', 'Wlochy', 'Australia', 'Francja', 'Polska', 'Portugalia', 'Rumunia', 'Austria', 'Czechy', 'Holandia', 'Szwajcaria', 'Slowacja', 'Belgia', 'Izrael', 'Szwecja'])
total_tests_per_country = np.array([115347287, 26679020, 18129900, 13689776, 12197500, 7935463, 11900095, 3599722, 2764867, 2627544, 1756856, 1515912, 2656731, 1470894, 528367, 3545600, 426328, 1789328])
tests_per_million = np.array([347922, 392435, 216200, 292768, 201821, 310224, 182201, 95483, 271371, 136837, 194765, 141483, 154954, 169617, 96764, 305393, 426328, 176874])
```

```python
sortowanie = np.argsort(total_tests_per_country)
total_tests_per_country1 = total_tests_per_country[sortowanie]
tests_per_million1 = tests_per_million[sortowanie]
countries1 = countries[sortowanie]
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15,14))



ax1.barh(countries1, total_tests_per_country1)
ax1.set_title("Liczba wszystkich testów", fontsize=23)
ax1.grid(axis="x", alpha=0.5)
ax1.set_axisbelow(True)
ax1.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))


xticks2 = np.arange(5)*10**5

ax2.barh(countries1, tests_per_million1)
ax2.set_title("Liczba testów na milion mieszkańców", fontsize=23)
ax2.grid(axis="x", alpha=0.5)
ax2.set_axisbelow(True)
ax2.set_xticks(xticks2)
ax2.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))



plt.show()
```
