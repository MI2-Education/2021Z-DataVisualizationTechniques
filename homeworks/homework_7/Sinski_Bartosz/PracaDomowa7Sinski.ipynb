## Poprawiona wizualizacja


```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import matplotlib.ticker as tic

studenci = [35.8, 2.1, 26.1, 13.4, 17, 5.2, 0.4]
ethnic = ["Asian", "Black", "Hispanic/Latino", "White","Non-Resident\n Alien","Ethnicity\n Unknown", "Other"]
pracownicy = [23.8, 3.3, 17, 44.9, 5.9, 4.5, 0.6]
fig, ax = plt.subplots(1,2,sharey=True,figsize=(20, 10))
ax[0].grid(zorder=0)
ax[1].grid(zorder=0)
ax[1].yaxis.set_tick_params(which='both', labelbottom=True)
ax[1].yaxis.set_major_formatter(tic.PercentFormatter())
ax[0].set_title("Ethnic diversity of Undergradute Students at Univesity of California-Irvine",loc="left")
ax[0].bar(ethnic,studenci,color="tab:gray",zorder=3)
ax[0].yaxis.set_major_formatter(tic.PercentFormatter())
ax[1].bar(ethnic,pracownicy,color="tab:gray",zorder=3)
ax[1].yaxis.set_major_formatter(tic.PercentFormatter())
ax[1].set_title("Ethnic diversity of Faculty at Univesity of California-Irvine",loc="left")

plt.show()
```
