```python
frame=pd.DataFrame(data={"col1":[0.783,0.737,0.704],"col2":["Male","National","Female"]})
plt.figure(figsize=(8, 6))

ax=sns.barplot(data=frame,x="col1",y="col2",color="b",dodge=False)
ax.set_axisbelow(True)
plt.grid(axis="x", which="both", zorder=0)
plt.ylabel("")
ax.set_xlabel('GNH', fontsize=18)
plt.title("Gross National Happines (GNH) index by Gender",fontsize=18)
plt.xticks(fontsize=17)
plt.yticks(fontsize=17)
for i, v in enumerate([0.783,0.737,0.704]):
    ax.text(v-0.11, i , str(v), color='white', fontweight='bold',size=17)
```
