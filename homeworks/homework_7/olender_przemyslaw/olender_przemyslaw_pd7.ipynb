### Mój poprawiony wykres

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('data.csv', parse_dates = ['date'], index_col='date', sep = ";")
```

```python
fig, ax = plt.subplots(figsize = (14, 10))

ax.plot(df.index.values, df['number'], linewidth = 3.0)


x_axis = np.array(['marzec', 'kwiecień', 'maj', 'czerwiec', 'lipiec', 'sierpień', 'wrzesień', 'październik'])
y_axis = np.array([5 * i * 10**4 for i in range(6)])

ax.set_yticklabels(['{:,.0f}'.format(i).replace(',', ' ') for i in y_axis], fontsize = 14)
ax.set_xticklabels(x_axis, fontsize = 14)

plt.grid()

plt.ylim(0, 320000)

plt.suptitle('Liczba osób objętych kwarantanną w Polsce', fontsize = 24)
plt.title('3 marca - 15 października 2020', fontsize = 18)
plt.ylabel('Liczba osób', fontsize = 18)
plt.show()
```
