## Kolejna poprawka

Jako poprawka tego drugiego wykresu stworzyłem za pomocą poniższego kodu następujący wykres:

```python
x = ["1.10", "2.10", "3.10", "4.10", "5.10", "6.10", "7.10", "8.10"]
y = [1967, 2292, 2367, 1934, 2006, 2236, 3003, 4280]

plt.rc("axes", axisbelow=True)

plt.figure(figsize=(12, 10))
plt.bar(x, y, color = "#1c57bd");
plt.title("Liczba nowych przypadków zakażenia koronawirusem w Polsce")
plt.xlabel("dzień")
plt.grid(axis="y", which="both", zorder=0)
plt.ylabel("liczba nowych przypadków")
plt.show()
```

## Komentarz
Uważam, że ten wykres jest lepszy od poprawionego, gdyż usunięta została linia "trendu", a dokładniej połączenie wierzchołków słupków, która nie dodawała żadnych nowych informacji, a jedynie zaciemniała obraz.

Poza tym, pozbyłem się etykiet danych, na rzecz czytelnej siatki w tle wykresu. Przy takim rzędzie wielkości, dokładne liczby przypadków są mało znaczące, a utrudniają odczyt wykresu. Pozbyłem się też siatki z osi x, gdyż ona również tylko zaciemniała wykres - słupki jednoznacznie wyznaczają dzień.

Z mniejszych detali, które jednak zacząłem zauważać, po komentarzu profesora Biecka - zadbałem o to, żeby słupki zaczynały się naprawdę w zerze - aby nie było odstępu między ramką, a początkiem osi OX. 

Wykres został również "ściśnięty" względem osi OX, to już wrażenie czysto estetyczne i subiektywne, ale moim zdaniem przerwy na poprzednim wykresie były zbyt duże.
