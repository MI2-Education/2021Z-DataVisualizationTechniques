## Zmiany wraz z uzasadnieniem

Uznałem, że na pierwszej poprawionej wersji wykresu obecny jest zbyt duży i nadmiarowy natłok informacji, stąd zdecydowałem się rozdzielić dane na dwa wykresy. Rozważałem po prostu usunięcie linii przedstawiającej trend, jednakże uznałem, że jest to na tyle ważna część wizualizacji, że nie można jej pominąć. Absolutnie zbyteczne były natomiast podpisy wartości, gdyż doskonale  można je odczytać bezpośrednio z wykresu.

Ze względów estetycznych zdecydowałem się też na usunięcie z barplota pionowych linii, gdyż powoduje jedynie zaciemnienie czytelności wykresu. W przpyadku linelpota siatka ta jest jednak niezbędna, aby wykres był czytelny. Zmieniłem, też kolorystykę wykresu na bardziej stonowaną, gdyż poprzedni wykres był zbyt krzykliwy.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
data={'Dzień' : ("01.10", "02.10", "03.10", "04.10", "05.10", "06.10", "07.10", "08.10"), 
      'Liczba Przypadków' : (1967, 2292, 2367, 1934, 2006, 2236, 3003, 4280)}
df=pd.DataFrame(data)
sns.set(rc={'figure.figsize':(14,14)})
sns.set_style("whitegrid")
pd.options.display.max_rows=10
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)
```

```python
plt.subplot(2,1,1)
sns.barplot(data = df, x = "Dzień", y="Liczba Przypadków", color="lightblue")
plt.title("Liczba nowych przypadków zakażenia koronawirusem w Polsce")
plt.ylim(0,4500)
plt.subplot(2,1,2)
sns.lineplot(data = df, x = "Dzień", y="Liczba Przypadków", color="red")
plt.title("Trend obrazujący liczbę nowych przypadków zakażenia koronawirusem w Polsce")
plt.ylim(0,4500)
plt.show()
```
