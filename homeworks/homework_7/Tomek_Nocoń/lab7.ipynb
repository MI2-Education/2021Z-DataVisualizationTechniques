```python
Ethnic = ["Hispanic", "American indain", "Asian", "Black", "White"]
Ethnic1 = []
Number = [1195, 408, 324, 1299, 531]
A = np.argsort(Number)[::-1]
for i in range(len(Number)):
    Ethnic1.append(Ethnic[A[i]])
Number.sort()
Number = Number[::-1]
```

```python
fig, ax = plt.subplots(figsize=(15,10))

bar_plot = plt.bar(Ethnic1, Number)

def label(bars):
    for idx,rect in enumerate(bar_plot):
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., height,
                Number[idx],
                ha='center', va='bottom', rotation=0)

autolabel(bar_plot)
plt.xlabel("Ethic")
plt.ylabel("Rate per 100,000 population")
plt.title('Race of COVID-19 cases by race/ethnic in Connecticut')
plt.show()
```
