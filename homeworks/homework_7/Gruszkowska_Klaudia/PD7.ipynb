## Moja poprawa

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.figure(figsize =(10,10))


sizes = [0.349, 54.814, 183.095, 166.973, 103.129, 103.913, 78.864, 95.641,
         86.712, 97.561,108.309,96.351, 71.311, 94.273, 143.442,159.530,296.020]
labels = ["04/03","21/03","2/04","11/04","23/04","7/05","26/05","16/06","10/07","30/07","14/08","31/08","4/09","16/09","26/09","6/10","15/10"]

p1 = sns.barplot(labels,sizes,color = 'dodgerblue')
p1.set(xlabel='Dni', ylabel='Liczba osób w tysiącach')
p1.set_xticklabels(p1.get_xticklabels(), rotation = -30)
plt.ylim([0, 330])


for idx, ax in np.ndenumerate(p1):
    for p in ax.patches:
        _x = p.get_x() + p.get_width() / 2 
        _y = p.get_y() + p.get_height() + 5
        value = int(p.get_height())
        ax.text(_x, _y, value, ha="center")

plt.title('Osoby na kwarantannie')
plt.show()
```
