## Drugie poprawienie wykresu z użyciem Pythona

```python
import matplotlib.pyplot as plt
import numpy as np

Metoda_płatności = ['Karta', 'Gotówka',
                      'Telefon,\nzegarek', 'Bankowość\ninternetowa', 'Inne']
procent = [66, 20, 7, 6, 1]

fig = plt.figure()
plt.bar(Metoda_płatności, procent, color="#33cc33")
plt.xlabel('Metoda płatności', fontsize=18)
plt.title('Czym najchętniej płacimy', fontsize=20)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.gca().set_yticklabels(['{:.0f}%'.format(i) for i in plt.gca().get_yticks()]) 
plt.rc('axes', axisbelow=True)
plt.grid(axis='y')
fig.set_figheight(7)
fig.set_figwidth(12)

plt.show()
```

## Dokonane zmiany

Wykres dalej pozostał wykresem słupkowym, który zdecydowanie czytelniej prezentuje zebrane dane. Dalej również pozostaje dodatkowa kategoria "Inne", żeby wszystkie wartości sumowały się do 100%. Natomiast zniknął gradient, który wydawało mi się, że będzie ułatwiał, a jednak utrudniał czytanie z wykresu i nic do niego nie wnosił. Dlatego też słupki pozostają tego samego koloru.
