## Wykres podwójnie poprawiony

```python
import matplotlib.pyplot as plt
import numpy as np
labels = ["Austria", "Germany", "Italy", "Poland", "Portugal", "United Kingdom"]
values = [1809263, 19215074, 12564713, 3521257, 2879577, 23709494]

fig = plt.figure(figsize=(15,9))
dane = list(zip(values,labels))
dane.sort()
plt.barh(*list(zip(*dane))[::-1], zorder=2)

plt.grid(zorder=0, axis="x")

ticks = np.arange(6)*5*10**6
plt.yticks(fontsize=12)
plt.xticks(ticks, ['{:,.0f}'.format(x).replace(',', ' ') for x in ticks], fontsize=12)
plt.title("COVID-19 tests done on 09.10.2020", fontsize=26)
plt.xlim(0, 27*10**6)
plt.xlabel("Number of tests", fontsize=18)

plt.show()
```
