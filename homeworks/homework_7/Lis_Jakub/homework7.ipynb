Wykres, który utworzyłem podczas drugiej pracy domowej:

Wykres poprawiony:

```python
years = np.arange(2007, 2021).tolist()
expenses = [50, 60.2, 64.8, 66.5, 69.2, 70.8, 74.8, 
            79, 79.8, 84.6, 90.4, 92.6, 97.6, 109]
```

```python
plt.figure(figsize=(14,7))
plt.bar(years, expenses, color="#17BECF", zorder=2)
ax  = plt.gca()
ax.set_xlim([2006.2, 2020.8])
ax.set_ylim([0, 125])
plt.xticks(years)
plt.title('\nPubliczne wydatki na służbę zdrowia na przestrzeni lat\n(w mld zł)\n',
          fontdict = {
              'fontsize': 20,
              'fontweight' : 500,
              'verticalalignment': 'baseline',
              'horizontalalignment': 'center'},
          loc='center')
for i in range(20, 121, 20):
    plt.axhline(i, color='black', linestyle='-', linewidth=0.3, zorder=1)
for i in range(len(years)):
    ax.text(years[i], expenses[i] + 2, str(expenses[i]), ha='center', fontsize=16)
plt.show()
```
