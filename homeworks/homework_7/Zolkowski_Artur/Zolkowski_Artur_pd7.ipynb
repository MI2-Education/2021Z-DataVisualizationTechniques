## Poprawa w Pythonie

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
sns.set(rc={'figure.figsize':(12,7)})
sns.set_style("darkgrid")
pd.options.display.max_rows=10
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)
```

```python
data = {"Emergency and operating":7760644,"Capital Project Reserve - Community Center":6581222,
        "CalPERS (Unfunded Liability)":5000000,"Projected Unassigned Fund Balance":3203032,
        "Total Nonspendable Fund Balances":1818322,  "Other Post Employment Benefits":1500000,
        "Technology reserve":912090,  
        }
s = pd.Series(data)
percent = s/s.sum() * 100
```

```python
sns.barplot(x=percent.index, y=percent.values,
           color = "grey")
plt.xticks(rotation=-20, ha="left")
plt.title("General Fund Reserves\nProjected Balance as of 9/30/2020")
plt.ylabel("%")
plt.show()
```
