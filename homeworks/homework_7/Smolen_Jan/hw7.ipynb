W drugiej pracy domowej zamieniłem pewne wykresy kołowe na wykresy słupkowe. Można jednak je dalej poprawić - pozbyć się zbędnych kolorów i legendy i dodać dokładne wartości nad słupkami.



```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
plt.figure(figsize=(16,8))
PercFac=[35.8, 2.1, 26.1,  13.4, 17, 5.2, 0.4]
Faculty=["Asian", "Black", "Hispanic/Latino", "White", "Non-Resident Alien",   "Ethnicity Unknown", "Other"]
ax=plt.bar(Faculty, PercFac)
plt.xticks(rotation='-45')
plt.ylim((0, 50))
plt.ylabel("Percentage of whole faculty")
plt.title("Ethnic diversity of Faculty at University of California - Irvine")
plt.style.use("seaborn-deep")
for p in ax.patches:
    percent='{:.1f}%'.format(p.get_height())
    x = p.get_x() +p.get_width()/2-0.1
    y = p.get_y() + p.get_height()+2
    plt.annotate(percent, (x, y))
```
