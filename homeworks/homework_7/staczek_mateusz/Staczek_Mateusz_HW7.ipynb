## Zmiany

Zmienić uznałem, że powinno się kolory z tęczowych na jednolite lub odróżniające źródła odnawialne od nieodnawialnych. Ponadto warto przesunąć słupki do krawędzi wykresu lub usunąć ramkę. Przy okazji usunąć można wartości przy słupkach.

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
from matplotlib.lines import Line2D

names = ["Węgiel kamienny","Węgiel brunatny","Paliwo jądrowe","Energia wiatru","Gaz ziemny",
         "Biomasa","Energia słoneczna","Energia wodna","Biogaz","Olej opałowy"]
values = [30,23,18,12,7,5,1,1,1,1]
```

```python
fig = plt.figure(figsize=(13,8))
ax = fig.add_subplot(1,1,1)
ax.xaxis.grid('on')
ax.set_frame_on(False)  
ax.barh(names[::-1],values[::-1],
        color=['red', 'green', 'green', 'green', 'green',
               'red', 'green', 'red', 'red', 'red'])

ax.xaxis.set_major_formatter(mtick.FormatStrFormatter('%.0f%%' ))
custom_lines = [Line2D([0], [0], color="red", lw=4),
                Line2D([0], [0], color="green", lw=4)]
ax.legend(custom_lines,('nieodnawialne', 'odnawialne'))
plt.title("Struktura wytwarzania energii elektrycznej w Polsce 2030r.")
plt.show()
```

#### Uzasadnienie

Kolory powinny być używane tylko tam, gdzie faktycznie coś znaczą, a ogólnie wykres powinien być bardziej minimalistyczny, ponieważ umieszczanie na nim zbędnych szczegółów (jak nie nie wnosząca ramka lub konkretne wartości przy słupkach, jeśli mamy wartości opisane przy osi) nie poprawia jego czytelności.
