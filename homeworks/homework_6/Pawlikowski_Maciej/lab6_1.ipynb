```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-3,3,20)
y=x**3+1
plt.plot(x,y)
plt.ylabel("y=x^3+1")
plt.xlabel("x")
plt.title("x^3+1", loc='center')
```

```python
fig=plt.figure()
fig, (ax1,ax2)=plt.subplots(1,2)
ax1.plot(x,y,color="orange")
ax1.set_ylabel("y=x^3+1")
ax1.set_xlabel("x")
ax1.set_title("x^3+1", loc='center')
ax2.plot(x,-x**4+30,label='-x^4+30')
ax2.plot(x,y,color="orange",label='x^3+1')
ax2.set_ylabel("y")
ax2.set_xlabel("x")
ax2.set_title("Title"=='center')
ax2.legend()
```
