# Zadania

### Zadanie 1

a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

### a)

```python
x = np.linspace(-3, 3, 100)
y = x ** 3 + 1
```

```python
plt.plot(x, y)
plt.ylabel('f(x)')
plt.xlabel('x')
plt.title('funkcja x^3 + 1 na przedziale [-3, 3]')
plt.show()
```

### b)

```python
z = -10*np.sqrt(abs(x))
```

```python
plt.plot(x, y, color = 'blue', label = 'x^3 - 1')
plt.plot(x, z, color = 'red', label = '-10*sqrt(|x|)', linewidth = 6.0)
plt.ylabel('y')
plt.xlabel('x')
plt.title('funkcje x^3 + 1 i -10*sqrt(|x|) na przedziale [-3, 3]')
plt.legend()
plt.show()
```

### c)

```python
fig, (fx, gx) = plt.subplots(1, 2, figsize=(15,5))
fx.plot(x, y)
fx.title.set_text('x^3 - 1')
fx.set_xlabel('x')
fx.set_ylabel('y')

gx.plot(x, z)
gx.title.set_text('-10*sqrt(|x|)')
gx.set_xlabel('x')
gx.set_ylabel('y')

```

# Zadanie 3

Poprawić wykres tak, aby łatwiej było odczytać w którym województwie śmiertelność wzrosła najbardziej.
Jest przynajmniej kilka innych podejść: scatterplot z latami na osi OX/OY, odcinki zaczynające się w danych ze średniej a kończące w 2020.
Jestem też ciekawa Waszych innych pomysłów.

![https://static.im-g.pl/im/7/26498/m26498997,WOJEWODZTWA2.png](https://static.im-g.pl/im/7/26498/m26498997,WOJEWODZTWA2.png)

Źródło: https://static.im-g.pl/im/7/26498/m26498997,WOJEWODZTWA2.png

```python
wojewodztwo = ["ZACHODNIOPOMORSKIE", "WIELKOPOLSKIE", "WARMINSKO-MAZURSKIE", "SWIETOKRZYSKIE",  "SLASKIE", "POMORSKIE", "PODLASKIE", "PODKARPACKIE", "OPOLSKIE", "MAZOWIECKIE", "MALOPOLSKIE", "LODZKIE", "LUBUSKIE", "LUBELSKIE", "KUJAWSKO-POMORSKIE", "DOLNOSLASKIE"]

l_2020 = [1874, 4013, 1565, 2008, 6270, 2521, 1468, 2909, 1356, 6564, 4743, 3687, 1313, 2731, 2435, 3665]

l_2015_2019 = [1509, 2796, 1241, 1208, 4255, 1830, 1049, 1643, 881, 4811, 2690, 2578, 894, 1920, 1768, 2683]
```

```python
N = len(wojewodztwo)

angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

values1 = l_2020
values1 += l_2020[:1]

values2 = l_2015_2019
values2 += values2[:1]
```

```python
import matplotlib.patches as mpatches
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.subplot(polar='True')

ax.scatter(angles, values1, s = 80, c = 'firebrick')  
ax.scatter(angles, values2, s = 70, c = 'dodgerblue')

#wstawienie nazw
plt.xticks(angles[:-1], wojewodztwo)

#pozycja podpisów osi
ax.set_rlabel_position(11)

#kolor osi
plt.yticks(color = 'grey')

#ręczne dodanie legendy
red = mpatches.Patch(color='firebrick', label = 'Liczba osób zmarłych w październiku 2020')
blue = mpatches.Patch(color='dodgerblue', label = 'Średnia liczba osób zmarłych w październiku w latach 2015-2019')
plt.legend(handles=[red, blue], loc = 'upper right', bbox_to_anchor=(1.3, 1.1))

plt.show()
```

Matplotlib pozwala na proste tworzenie wykresów w stylu xkcd, służy do tego funkcja `xkcd()` z `matplotlib.pyplot`.
Warto zajrzeć do dokumentacji: https://matplotlib.org/gallery/showcase/xkcd.html
Mój przykład umieściłam poniżej.

Uwaga, w szczególności styl działa też z wykresami kołowymi i słupkowymi.

### Przykład

```python
x = [0, 0.05, 0.06, 0.9, 1]
y = [0, 0.5, 0, 0, 4]

# rysowanie w xkcd mode
with plt.xkcd():
    # ustawienie rozmiaru
    plt.figure(figsize=(10,10))
    
    plt.plot(x,y)
    
    # usuniecie podpisow osi
    plt.xticks([])
    plt.yticks([])
    
    # dodanie napisow
    plt.title("TWD project")
    plt.xlabel('Time')
    plt.ylabel('Productivity')

    plt.annotate(
        'One day untli deadline',
        xy=(0.9, 0), arrowprops=dict(arrowstyle='->'), xytext=(0.5, 1))
plt.show()
```

```python
data = [6,10,84]
names = ['making chart', 'specifing idea', 'wondering what to do']
```

```python
with plt.xkcd():
    plt.pie(data, labels = names)
    plt.title("Time spent on exercise 3, HW7, TWD 2020")
plt.show()
```

```python
# i jeszcze barplot, bo kołowych nie lubimy

with plt.xkcd():
    plt.figure(figsize=(10,5))
    
    plt.barh(names, data)

    plt.title("Time spent on exercise 3, HW7, TWD 2020")
    plt.ylabel('Activity')
    plt.xlabel('percent of time')

plt.show()
```
