### Zadanie 1

a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

```python
x = np.linspace(-3,3,100)
y = x**3+1
plt.plot(x,y)
plt.title("Wykres")
plt.ylabel("oś Y")
plt.xlabel("oś X")
plt.show()
```

```python
y_2 = x**2 - x
plt.plot(x,y,label="x^3+1")
plt.plot(x,y_2,linewidth = 5,color = 'red',label="x^2 - x")
plt.title("Wykres 2")
plt.ylabel("oś Y")
plt.xlabel("oś X")
plt.legend()
plt.show()
```

```python
fig, wykresy = plt.subplots(2,1)
wykresy[0].plot(x,y)
wykresy[0].set_ylabel("oś y")
wykresy[0].set_xlabel("oś x")
wykresy[1].plot(x,y,label="x^3+1")
wykresy[1].plot(x,y_2,linewidth = 5,color = 'red',label="x^2 - x")
wykresy[1].set_xlabel("oś x")
wykresy[1].set_ylabel("oś y")
wykresy[1].legend()
fig.suptitle("Dwa wykresy")
plt.show()
```

# Zadanie 3

Poprawić wykres tak, aby łatwiej było odczytać w którym województwie śmiertelność wzrosła najbardziej.
Jest przynajmniej kilka innych podejść: scatterplot z latami na osi OX/OY, odcinki zaczynające się w danych ze średniej a kończące w 2020.
Jestem też ciekawa Waszych innych pomysłów.

![https://static.im-g.pl/im/7/26498/m26498997,WOJEWODZTWA2.png](https://static.im-g.pl/im/7/26498/m26498997,WOJEWODZTWA2.png)

Źródło: https://static.im-g.pl/im/7/26498/m26498997,WOJEWODZTWA2.png

```python
import matplotlib.lines as mlines
wojewodztwo = ["ZACHODNIOPOMORSKIE", "WIELKOPOLSKIE", "WARMINSKO-MAZURSKIE", "SWIETOKRZYSKIE",  "SLASKIE", "POMORSKIE", "PODLASKIE", "PODKARPACKIE", "OPOLSKIE", "MAZOWIECKIE", "MALOPOLSKIE", "LODZKIE", "LUBUSKIE", "LUBELSKIE", "KUJAWSKO-POMORSKIE", "DOLNOSLASKIE"]
l_2020 = [1874, 4013, 1565, 2008, 6270, 2521, 1468, 2909, 1356, 6564, 4743, 3687, 1313, 2731, 2435, 3665]
l_2015_2019 = [1509, 2796, 1241, 1208, 4255, 1830, 1049, 1643, 881, 4811, 2690, 2578, 894, 1920, 1768, 2683]
plt.figure(figsize=(10,10))
w2 =plt.scatter(wojewodztwo,l_2020,color = 'tab:blue',label="Liczba osób zmarłych w październiku 2020",s=100)
w1 = plt.scatter(wojewodztwo,l_2015_2019,color = 'tab:red',label="Średnia liczba osób zmarłych w październiku w latach 2015-2019",s=100)
for i in range(16):
    ytextu = (l_2020[i] + l_2015_2019[i])/2
    text = l_2020[i] - l_2015_2019[i]
    plt.text(wojewodztwo[i],ytextu,text)
    plt.plot([wojewodztwo[i],wojewodztwo[i]],[l_2015_2019[i],l_2020[i]],'--',color = 'black', markersize = 0,linewidth = 1.5)

linepatch = mlines.Line2D([], [],color = 'black',ls = '--',label='Rożnica liczby osób zmarłych pomiędzy porównywanymi okresami')   
plt.legend(handles=[w2,w1,linepatch],bbox_to_anchor=(1, 1))
plt.xticks(rotation=90 )
plt.ylabel("Liczba osób zmarłych")
plt.title('Porównywanie liczby osób zmarłych w poszczególnych województwach\nw październiku 2020 i średnio w latach 2015-2019',loc='left')
plt.grid(True)
plt.show()
    
```

Matplotlib pozwala na proste tworzenie wykresów w stylu xkcd, służy do tego funkcja `xkcd()` z `matplotlib.pyplot`.
Warto zajrzeć do dokumentacji: https://matplotlib.org/gallery/showcase/xkcd.html
Mój przykład umieściłam poniżej.

Uwaga, w szczególności styl działa też z wykresami kołowymi i słupkowymi.

### Przykład

```python
x = [0, 0.05, 0.06, 0.9, 1]
y = [0, 0.5, 0, 0, 4]

# rysowanie w xkcd mode
with plt.xkcd():
    # ustawienie rozmiaru
    plt.figure(figsize=(10,10))
    
    plt.plot(x,y)
    
    # usuniecie podpisow osi
    plt.xticks([])
    plt.yticks([])
    
    # dodanie napisow
    plt.title("TWD project")
    plt.xlabel('Time')
    plt.ylabel('Productivity')

    plt.annotate(
        'One day untli deadline',
        xy=(0.9, 0), arrowprops=dict(arrowstyle='->'), xytext=(0.5, 1))
plt.show()
```

```python
narzedzia = ["Canva","PowerPoint","LucidPress","Overleaf"]
wartosci = [14,2,2,0]
with plt.xkcd():
    plt.figure(figsize=(10,10))
    plt.bar(narzedzia,wartosci,color="grey")
    plt.title("Liczba zespołów korzystających z narzędzi zaproponowanych\n przez prowadzącego do tworzenia plakatu na pierwszy projekt")
plt.show()
```
