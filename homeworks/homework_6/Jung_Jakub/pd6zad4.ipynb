```python
import matplotlib.pyplot as plt

x = [0, 6, 10]
y = [0, 6, 6]


with plt.xkcd():
    # ustawienie rozmiaru
    plt.figure(figsize=(10,10))
    
    plt.plot(x,y, color = 'blue')
    plt.ylim(0,10)

    
    # usuniecie podpisow osi
    plt.xticks([])
    plt.yticks([])
    
    # dodanie napisow
    plt.title("Aktywność na studiach")
    plt.xlabel('Czas')
    plt.ylabel('Punkty za aktywność')

    plt.annotate(
        'Wydanie Cyberpunk 2077',
        xy=(6, 6), arrowprops=dict(arrowstyle='->'), xytext=(4, 7))
plt.show()
```
