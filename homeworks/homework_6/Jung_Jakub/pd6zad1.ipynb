```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(-3, 3, 1000)
y1 = x**3+1
y2 = np.sin(4*x)
plt.plot(x, y1)
plt.plot(x, y2, color = 'red', linewidth = 3)
plt.ylim([-30, 30])
plt.legend(["x^3+1", "sin(4x)"])
plt.title("Różne funkcje")
plt.ylabel("y")
plt.xlabel("x")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x, y1)
ax1.set_ylim([-30, 30])
ax2.plot(x, y2, color = 'red')
ax2.set_ylim([-30, 30])
plt.show()
```
