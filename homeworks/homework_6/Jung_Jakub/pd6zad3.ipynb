```python
import matplotlib.pylab as plt
import numpy as np
import pandas as pd

wojewodztwa = ["Zachodniopomorskie",
              "Wielkopolskie",
              "Warmińsko-Mazurskie",
              "Świętokrzyskie",
              "Śląskie",
              "Pomorskie",
              "Podlaskie",
              "Podkarpackie",
              "Opolskie",
              "Mazowieckie",
              "Małopolskie",
              "Łódzkie",
              "Lubuskie",
              "Lubelskie",
              "Kujawsko-Pomorskie",
              "Dolnośląskie"]
zmarli_kiedys = [1509, 2796, 1241, 1208, 4255, 1830, 1049, 1643, 881, 4811, 2690, 2578, 894, 1930, 1768, 2683]
zmarli_2020 = [1874, 4013, 1565, 2008, 6270, 2521, 1478, 2909, 1356, 6564, 4743, 3687, 1313, 2731, 2435, 3665]
wzrost = []
for i in range(len(zmarli_2020)):
    wzrost.append(zmarli_2020[i]/zmarli_kiedys[i]*100 - 100)
    
df = pd.DataFrame({"Wojewodztwo":wojewodztwa, "Wzrost":wzrost}).sort_values('Wzrost',ascending=True)
plt.barh("Wojewodztwo", "Wzrost", data=df)
plt.xlim([0, 100])
plt.xlabel("Wzrost zmarłych (%)")
plt.ylabel("Województwo")
plt.title("Wzrost liczby zmarłych w październiku 2020 roku w porównaniu do lat 2015-2019")
plt.sort(wzrost)
plt.show()
```
