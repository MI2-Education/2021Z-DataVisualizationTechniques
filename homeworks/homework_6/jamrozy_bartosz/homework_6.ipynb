### Zadanie 1

a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

```python
import numpy as np
x=np.linspace(-3,3,100)
y=x**3+1
plt.plot(x,y,label="y=x^3+1")
plt.title("Wykres a i b ")
plt.xlabel("x")
plt.ylabel("y")
y1=x**2
plt.plot(x,y1,label="y=x^2",color="red",linewidth=2)
plt.show()

fig, (ax1,ax2) = plt.subplots(1,2)
fig.suptitle("Wykres a i b ")
ax1.plot(x,y,label="y=x^3+1")
ax1.set_title("y=x^3+1")
ax2.plot(x,y1,label="y=x^2",color="red",linewidth=2)
ax2.set_title("y=x^2")
plt.show()
```

Matplotlib pozwala na proste tworzenie wykresów w stylu xkcd, służy do tego funkcja `xkcd()` z `matplotlib.pyplot`.
Warto zajrzeć do dokumentacji: https://matplotlib.org/gallery/showcase/xkcd.html
Mój przykład umieściłam poniżej.

Uwaga, w szczególności styl działa też z wykresami kołowymi i słupkowymi.

### Przykład

```python
x = [0, 0.05, 0.06, 0.9, 1]
y = [0, 0.5, 0, 0, 4]

# rysowanie w xkcd mode
with plt.xkcd():
    # ustawienie rozmiaru
    plt.figure(figsize=(10,10))
    
    plt.plot(x,y)
    
    # usuniecie podpisow osi
    plt.xticks([])
    plt.yticks([])
    
    # dodanie napisow
    plt.title("TWD project")
    plt.xlabel('Time')
    plt.ylabel('Productivity')

    plt.annotate(
        'One day untli deadline',
        xy=(0.9, 0), arrowprops=dict(arrowstyle='->'), xytext=(0.5, 1))
plt.show()
```

```python
labels=["Nieudane","złe"]
y=[100,1]
with plt.xkcd():
    plt.pie(y,labels=labels)
    plt.title("Jakie są wykresy kołowe")
    plt.annotate(
        '99%',
        xy=(-0.5, -0.5), arrowprops=dict(arrowstyle='->'), xytext=(-1,-1))
    plt.annotate(
        '2%',
        xy=(1, 0), arrowprops=dict(arrowstyle='->'), xytext=(1, 0.5))
plt.show()    
```
