### Zadanie 1

a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

```python
# a) b)
x = np.linspace(-3, 3, 100)
y = x**3 + 1
y2 = np.log(np.abs(x))
plt.plot(x, y, label = "f(x)")
plt.title("f(x) = x^3 + 1 i g(x) = log(x)", loc = "center")
plt.plot(x, y2, label = "g(x)", linestyle = "--", color = "red", linewidth = 3)
plt.legend()
plt.show()
```

```python
# c)
fig, (ax1, ax2) = plt.subplots(1, 2) # 2 wiersze, 1 kolumna
fig.suptitle('Subplot')

ax1.plot(x, y, label = "f(x)")
ax2.plot(x, y2, label = "g(x)", linestyle = "--", color = "red", linewidth = 3)
fig.suptitle("Wykresy")
plt.show()
```

```python
labels = ['Blue', 'Red', 'Green', 'Purple']
sizes = [15, 30, 45, 10]

plt.bar(labels, sizes)
plt.show()
```

Matplotlib pozwala na proste tworzenie wykresów w stylu xkcd, służy do tego funkcja `xkcd()` z `matplotlib.pyplot`.
Warto zajrzeć do dokumentacji: https://matplotlib.org/gallery/showcase/xkcd.html
Mój przykład umieściłam poniżej.

Uwaga, w szczególności styl działa też z wykresami kołowymi i słupkowymi.

### Przykład

```python
x = [0, 0.05, 0.06, 0.9, 1]
y = [0, 0.5, 0, 0, 4]

# rysowanie w xkcd mode
with plt.xkcd():
    # ustawienie rozmiaru
    plt.figure(figsize=(10,10))
    
    plt.plot(x,y)
    
    # usuniecie podpisow osi
    plt.xticks([])
    plt.yticks([])
    
    # dodanie napisow
    plt.title("TWD project")
    plt.xlabel('Time')
    plt.ylabel('Productivity')

    plt.annotate(
        'One day untli deadline',
        xy=(0.9, 0), arrowprops=dict(arrowstyle='->'), xytext=(0.5, 1))
plt.show()
```

```python
x = np.linspace(-2, 2, 100)
x2 = np.linspace(-5, 5, 100)
up = np.sqrt(1-(np.abs(x)-1)**2) + np.pi
down = np.arccos(1-np.abs(x))
other = x2 * 0 + 0.2

# rysowanie w xkcd mode
with plt.xkcd():
    # ustawienie rozmiaru
    plt.figure(figsize=(10,5))
    
    plt.plot(x, up, color = "blue")
    plt.plot(x, down, color = "blue")
    plt.plot(x2, other, color = "blue")
    
    # usuniecie podpisow osi
    plt.xticks([])
    plt.yticks([])
    
    # dodanie napisow
    plt.title("Function showing how much I love StackOverflow in time")
    plt.xlabel('Time')
    plt.ylabel('How much I love Stack')

    plt.annotate(
        'First day at MINI',
        xy=(0, 0.2), arrowprops=dict(arrowstyle='->'), xytext=(1, 1))
plt.show()
```
