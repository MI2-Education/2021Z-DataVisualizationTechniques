### Zadanie 1

a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

## a i b

x = np.linspace(-3, 3, 1000)
y_1 = x**3 + 1
plt.plot(x, y_1, label = "y = x^3+1")
y_2 = np.sin(x)
plt.plot(x, y_2, label = "y = sin(x)", color = "green", linewidth = 4)
plt.legend()
plt.title("Wykresy funkcji")
plt.ylabel("y")
plt.xlabel("x")
plt.show()

## c
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.plot(x, y_1)
ax1.set_title("y = x^3+1")
ax1.set_ylabel("y")
ax1.set_xlabel("x")

ax2 = fig.add_subplot(1,2,2)
ax2.plot(x, y_2)
ax2.title.set_text("y = sin(x)")
ax2.set_ylabel("y")
ax2.set_xlabel("x")


plt.subplots_adjust(wspace = 0.5)
plt.show()
```

Poprawić wykres tak, aby łatwiej było odczytać w którym województwie śmiertelność wzrosła najbardziej.

```python
wojewodztwo = ["ZACHODNIOPOMORSKIE", "WIELKOPOLSKIE", "WARMINSKO-MAZURSKIE", "SWIETOKRZYSKIE",  "SLASKIE", "POMORSKIE", "PODLASKIE", "PODKARPACKIE", "OPOLSKIE", "MAZOWIECKIE", "MALOPOLSKIE", "LODZKIE", "LUBUSKIE", "LUBELSKIE", "KUJAWSKO-POMORSKIE", "DOLNOSLASKIE"]

l_2020 = [1874, 4013, 1565, 2008, 6270, 2521, 1468, 2909, 1356, 6564, 4743, 3687, 1313, 2731, 2435, 3665]

l_2015_2019 = [1509, 2796, 1241, 1208, 4255, 1830, 1049, 1643, 881, 4811, 2690, 2578, 894, 1920, 1768, 2683]


df = pd.DataFrame(zip(wojewodztwo, l_2020, l_2015_2019), columns = ['wojewodztwo' , 'l_2020', 'l_2015_2019'])
df["o_ile_procent_wzoroslo_wzgledem_sredniej"] = (df["l_2020"] - df["l_2015_2019"])/df["l_2015_2019"] * 100
df = df.sort_values(by = "o_ile_procent_wzoroslo_wzgledem_sredniej")


fig, ax = plt.subplots(figsize=(10,6))

ax.barh(df['wojewodztwo'],df['o_ile_procent_wzoroslo_wzgledem_sredniej'])

vals = ax.get_xticks()
ax.set_xticklabels(['{:,.0f}%'.format(x) for x in vals])
ax.set_title("O ile procent wzrosła liczba zgonów w 2020 roku względem średniej z lat 2015-2019\nz podziałem na województwa")
ax.set_xlabel("Procentowy wzrost")
ax.grid(axis = 'x')
plt.show()
```

Matplotlib pozwala na proste tworzenie wykresów w stylu xkcd, służy do tego funkcja `xkcd()` z `matplotlib.pyplot`.
Warto zajrzeć do dokumentacji: https://matplotlib.org/gallery/showcase/xkcd.html
Mój przykład umieściłam poniżej.

Uwaga, w szczególności styl działa też z wykresami kołowymi i słupkowymi.

### Przykład

```python
x = [0, 0.05, 0.06, 0.9, 1]
y = [0, 0.5, 0, 0, 4]

# rysowanie w xkcd mode
with plt.xkcd():
    # ustawienie rozmiaru
    plt.figure(figsize=(10,10))
    
    plt.plot(x,y)
    
    # usuniecie podpisow osi
    plt.xticks([])
    plt.yticks([])
    
    # dodanie napisow
    plt.title("TWD project")
    plt.xlabel('Time')
    plt.ylabel('Productivity')

    plt.annotate(
        'One day untli deadline',
        xy=(0.9, 0), arrowprops=dict(arrowstyle='->'), xytext=(0.5, 1))
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import gridspec

custpref=pd.DataFrame({'tov_type':['Bad','Also bad but in white'],'count':[24,76]})


with plt.xkcd():

    fig=plt.figure(figsize=(6,4))
    gs1 = gridspec.GridSpec(1,1,
        left=0.1,right=0.7,
        bottom=0.1,top=0.7,
    )
    gs2 = gridspec.GridSpec(1,1,
        left=0.05,right=0.95,
        bottom=0.9,top=1.0,
    )

    pie_ax=fig.add_subplot(gs1[0])

    colors = ["black","white"]


    wedges, texts = pie_ax.pie(

        custpref['count'],

        shadow=False,

    colors=colors,

        startangle=120, wedgeprops={"edgecolor":"k",'linewidth': 5, 'linestyle': 'solid', 'antialiased': True}
    )

    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    kw = dict(xycoords='data', textcoords='data', arrowprops=dict(arrowstyle="-"), zorder=0, va="center")
    
    colors = ["black","black"]
    
    for i, p in enumerate(wedges):
        ang = (p.theta2 - p.theta1)/2. + p.theta1
        y = np.sin(np.deg2rad(ang))
        x = np.cos(np.deg2rad(ang))
        horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
        connectionstyle = "angle,angleA=0,angleB={}".format(ang)
        kw["arrowprops"].update({"connectionstyle": connectionstyle,"color":colors[i]})
        pie_ax.annotate(custpref['tov_type'][i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                     horizontalalignment=horizontalalignment, **kw)


pie_ax.set_title("How good are piecharts?")
plt.show()
```
