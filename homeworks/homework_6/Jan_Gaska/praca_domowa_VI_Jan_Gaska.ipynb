## Zadanie 1

a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

```python
x = np.linspace(-3,3,100)
y = x*x*x +1
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y=x^3+1",loc = "center")
plt.show()
```

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

```python
y_2 = np.sin(2*x+4)
plt.plot(x,y,label = "y=x^3+1")
plt.plot(x,y_2,label = "y=sin(2x+4)",color="green", linewidth = 7)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Dwa wykresy")
plt.legend()
plt.show
```

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

```python
fig, (ax1,ax2) = plt.subplots(1,2)
fig.suptitle("Wykresy:")
ax1.plot(x,y)
ax1.set_title("y=x^3+1")
ax2.plot(x,y_2,color="green", linewidth = 7)
ax2.set_title("y=sin(2x+4)")
plt.show()
```

# Zadanie 3

Poprawić wykres tak, aby łatwiej było odczytać w którym województwie śmiertelność wzrosła najbardziej. Jest przynajmniej kilka innych podejść: scatterplot z latami na osi OX/OY, odcinki zaczynające się w danych ze średniej a kończące w 2020. Jestem też ciekawa Waszych innych pomysłów.

```python
wojewodztwo = np.array(["ZACHODNIOPOMORSKIE", "WIELKOPOLSKIE", "WARMINSKO-MAZURSKIE", "SWIETOKRZYSKIE",  "SLASKIE", "POMORSKIE", "PODLASKIE", "PODKARPACKIE", "OPOLSKIE", "MAZOWIECKIE", "MALOPOLSKIE", "LODZKIE", "LUBUSKIE", "LUBELSKIE", "KUJAWSKO-POMORSKIE", "DOLNOSLASKIE"])
l_2020 = np.array([1874, 4013, 1565, 2008, 6270, 2521, 1468, 2909, 1356, 6564, 4743, 3687, 1313, 2731, 2435, 3665])
l_2015_2019 = np.array([1509, 2796, 1241, 1208, 4255, 1830, 1049, 1643, 881, 4811, 2690, 2578, 894, 1920, 1768, 2683])
roznica_bezwzgledna = l_2020 - l_2015_2019
roznica_wzgledna = 100*(l_2020-l_2015_2019)/l_2015_2019


fig, (ax1,ax2) = plt.subplots(2,1)
fig.set_size_inches(18.5, 10.5)
fig.suptitle("Wzrost smiertelnosci:")
ax1.barh(wojewodztwo,roznica_wzgledna)
ax1.set_title("Wzgledny")
ax2.barh(wojewodztwo,roznica_bezwzgledna)
ax2.set_title("Bezwzgledny")
plt.show()
```

Matplotlib pozwala na proste tworzenie wykresów w stylu xkcd, służy do tego funkcja xkcd() z matplotlib.pyplot. Warto zajrzeć do dokumentacji: https://matplotlib.org/gallery/showcase/xkcd.html Mój przykład umieściłam poniżej.

Uwaga, w szczególności styl działa też z wykresami kołowymi i słupkowymi.

### Przykład

```python
x = [0, 0.05, 0.06, 0.9, 1]
y = [0, 0.5, 0, 0, 4]

# rysowanie w xkcd mode
with plt.xkcd():
    # ustawienie rozmiaru
    plt.figure(figsize=(10,10))
    
    plt.plot(x,y)
    
    # usuniecie podpisow osi
    plt.xticks([])
    plt.yticks([])
    
    # dodanie napisow
    plt.title("TWD project")
    plt.xlabel('Time')
    plt.ylabel('Productivity')

    plt.annotate(
        'One day untli deadline',
        xy=(0.9, 0), arrowprops=dict(arrowstyle='->'), xytext=(0.5, 1))
plt.show()
```

```python
x = [20,5,55]
with plt.xkcd():
    plt.figure(figsize=(20,20))
    plt.pie(x, startangle=-140, colors=["yellow","grey","lightblue"])
    plt.title("A pyramid", size=35, loc="center")
    plt.text(0,1.1,'Sky', fontsize=25)
    plt.text(0,-1.1,"Pyramid", fontsize=20)
    plt.text(0.9,-0.6,"Shadow", fontsize=20)
plt.show()
```
