### Zadanie 1
a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

```python
import numpy as np
x = np.linspace(-3,3,10)
y = x**3+1
plt.plot(x,y)
plt.title("Wykres y", loc = "center")
plt.ylabel('y')
plt.xlabel('x')
plt.show()
```

```python
x = np.linspace(-3,3,10)
y = x**3+1
y2 = x**2+4
plt.plot(x, y, label = "y = x^3 + 1")
plt.plot(x, y2, label = "y = x^2 + 4", linewidth = 5 )
plt.title("Wykres y", loc = "center")
plt.ylabel('y')
plt.xlabel('x')
plt.legend()
plt.show()
```

```python
fig, (plt1, plt2) = plt.subplots(1,2)
plt1.plot(x,y)
plt1.set(xlabel = 'x',ylabel = 'y')



plt2.plot(x, y2, label = "y = x^2 + 4", linewidth = 5 )
plt1.set(xlabel = 'x',ylabel = 'y')
plt2.legend()
```

Matplotlib pozwala na proste tworzenie wykresów w stylu xkcd, służy do tego funkcja xkcd() z matplotlib.pyplot. Warto zajrzeć do dokumentacji: https://matplotlib.org/gallery/showcase/xkcd.html Mój przykład umieściłam poniżej.

Uwaga, w szczególności styl działa też z wykresami kołowymi i słupkowymi.

### Przykład

```python
x = [0, 0.05, 0.06, 0.9, 1]
y = [0, 0.5, 0, 0, 4]

# rysowanie w xkcd mode
with plt.xkcd():
    # ustawienie rozmiaru
    plt.figure(figsize=(10,10))
    
    plt.plot(x,y)
    
    # usuniecie podpisow osi
    plt.xticks([])
    plt.yticks([])
    
    # dodanie napisow
    plt.title("TWD project")
    plt.xlabel('Time')
    plt.ylabel('Productivity')

    plt.annotate(
        'One day untli deadline',
        xy=(0.9, 0), arrowprops=dict(arrowstyle='->'), xytext=(0.5, 1))
plt.show()
```

```python
sizes = [90,10]
labels = ["No","No but in yellow"]

with plt.xkcd():
    plt.figure(figsize=(10,10))
    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, labels=labels, startangle=70, colors = ["blue","yellow"])
    ax1.axis('equal') 
    plt.title("Should you use pie charts?") 
plt.show()
```
