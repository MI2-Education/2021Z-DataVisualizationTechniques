### Zadanie 1

a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

```python
#a)
x = np.linspace(-3,3,num=100)
y = x**3 + 1
plt.plot(x,y,label="y=x^3+1")
plt.title("y=x^3+1")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```python
#b)
plt.plot(x,y,label="y=x^3+1")
plt.plot(x,-(y)+6,"r:",label="y=-x^3+7")
plt.title("Wykresy funkcji")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
#c)
fig, ax = plt.subplots(2,sharex=True,sharey=True)
ax[0].plot(x,y,label="y=x^3+1")
ax[1].plot(x,-y+6,"r:",label="y=-x^3+7")
fig.suptitle("Wykresy funkcji")
plt.xlabel("x")
plt.ylabel("y")
fig.show()
```

Matplotlib pozwala na proste tworzenie wykresów w stylu xkcd, służy do tego funkcja `xkcd()` z `matplotlib.pyplot`.
Warto zajrzeć do dokumentacji: https://matplotlib.org/gallery/showcase/xkcd.html
Mój przykład umieściłam poniżej.

Uwaga, w szczególności styl działa też z wykresami kołowymi i słupkowymi.

### Przykład

```python
x = [0,1]
y = [5/6,1/6]

# rysowanie w xkcd mode
with plt.xkcd():
    # ustawienie rozmiaru
    # plt.figure(figsize=(10,10))
    
    plt.pie(y,colors=["yellow","white"],startangle=23)
    
    # plt.xticks([])
    # plt.yticks([])
    

    plt.title("Distribution of a pacman in a pacman")
    # plt.xlabel('Time')
    plt.ylabel('')

    plt.annotate(
        'Not pacman',
        xy=(0.9, 0), arrowprops=dict(arrowstyle='->'), xytext=(1.2, 0.3))
    plt.annotate(
        'Pacman',
        xy=(-1, 0), arrowprops=dict(arrowstyle='->'), xytext=(-2.3, -0.3))
    plt.text(0.2,0.39,"O")
    plt.annotate(
        'The letter O',
        xy=(0, 0.50), arrowprops=dict(arrowstyle='->'), xytext=(-2.3, 0.6))
plt.show()
```
