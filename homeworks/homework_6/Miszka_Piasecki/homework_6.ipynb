### Zadanie 1

a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

```python
def f(x):
    return pow(x,3) + 1

x = np.linspace(-3, 3)
y = pow(x,3) + 1
z = 10*np.sin(x)
plt.plot(x, y, label = "x^3 + 1", color = "cornflowerblue")
plt.plot(x, z, label = "10*sin(x)", color = "darkorange")
plt.ylabel("Y axis")
plt.xlabel("X Axis")
plt.title("Bardzo ładnie to wygląda :)", loc = "center")
plt.legend(loc = "lower right")
plt.show()
fig, (ax1, ax2) = plt.subplots(2,1)
ax1.plot(x,y, color = "black")
ax2.plot(x,z, color = "red")

```

```python
srednia_poprzednie = [1509,2796, 1241, 1208, 4255, 1830, 1049, 1643, 881, 4811, 2690, 2578, 894, 1930, 1768, 2683]
rok_2020 = [1874, 4013, 1565, 2008, 6270, 2521, 1478, 2909, 1356, 6564, 4743, 3687, 1313, 2731, 2435, 3665]
wojewodztwa = ["zachodniopomorskie", "wielkopolskie","warmińsko-mazurskie", "świętokrzyskie","śląskie","pomorskie","podlaskie","podkarpackie","opolskie","mazowieckie","małopolskie","łódzkie","lubuskie","lubelskie","kujawsko-pomorskie","dolnośląskie"]
roznica = np.subtract(rok_2020, srednia_poprzednie)
tablica = {'Wojewodztwa':wojewodztwa,'Roznica':roznica}
tablica = pd.DataFrame(tablica)
tablica = tablica.sort_values("Roznica", ascending = False)
plt.barh(tablica["Wojewodztwa"],tablica["Roznica"], color = "red")
plt.title("Przyrost zgonów w październiku 2020 w porównaniu do poprzednich lat")
plt.xlabel("Przyrost zgonów")
plt.ylabel("Województwa")
```

Matplotlib pozwala na proste tworzenie wykresów w stylu xkcd, służy do tego funkcja `xkcd()` z `matplotlib.pyplot`.
Warto zajrzeć do dokumentacji: https://matplotlib.org/gallery/showcase/xkcd.html
Mój przykład umieściłam poniżej.

Uwaga, w szczególności styl działa też z wykresami kołowymi i słupkowymi.

### Przykład

```python
x = ["Jan","Feb","March","Apr","May","June","July","Aug", "Sept","Oct","Nov","Dec"]
y = [1.5, 1.6,  1.7, 1.6, 1.1, 1.2, 1.5,1.7,1.65,1.4,1.2,1]


# rysowanie w xkcd mode
with plt.xkcd():
    # ustawienie rozmiaru
    plt.figure(figsize=(10,7.5))
    
    plt.plot(x,y)
    
    # usuniecie podpisow osi
    
    plt.yticks([])
    
    # dodanie napisow
    plt.title("2020")
    plt.xlabel('Time')
    plt.ylabel('Willingness to live')
    plt.annotate(
        'One day untli deadline',
        xy=(2, 2), arrowprops=dict(arrowstyle='->'), xytext=(0.2, 1))

    
plt.show()
```
