### Zadanie 1

a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

```python
x = np.linspace(-3, 3, 100)
y = x**3 + 1
plt.plot(x, y, label = "f(x) = x^3 + 1")
#plt.ylabel()
plt.title("Wykresy", loc = 'center')
plt.ylabel('f(x)')
plt.xlabel('x')
y1 = x**7 - 14*(x**5) + 49 * (x**3) - 36*x
plt.plot(x, y1, color = "red", linestyle = "--", label = "f(x) = x^7 - 14*x^5 + 49*x^3 - 36*x", )
plt.legend()
plt.show()

fig, (ax1, ax2) = plt.subplots(2, 1)

ax1.plot(x,y)
ax2.plot(x,y1, color = "red", linestyle = "--")
plt.show()
```

Poprawić wykres tak, aby łatwiej było odczytać w którym województwie śmiertelność wzrosła najbardziej.
Jest przynajmniej kilka innych podejść: scatterplot z latami na osi OX/OY, odcinki zaczynające się w danych ze średniej a kończące w 2020.
Jestem też ciekawa Waszych innych pomysłów.

![https://static.im-g.pl/im/7/26498/m26498997,WOJEWODZTWA2.png](https://static.im-g.pl/im/7/26498/m26498997,WOJEWODZTWA2.png)

Źródło: https://static.im-g.pl/im/7/26498/m26498997,WOJEWODZTWA2.png

```python
zmarli_2020 = [1874, 4013, 1565, 2008, 6270, 2521, 1478, 2909, 1356, 6564, 4743, 3687, 1313, 2731, 2435, 3665]
srednia = [1509, 2796, 1241, 1208, 4255, 1830, 1049, 1643,881, 4811, 2690, 2578, 894, 1930, 1768, 2683]
names = ['ZACHODNIO-POMORSKIE', 'WIELKOPOLSKIE','WARMIŃSKO-MAZURSKIE','ŚWIĘTOKRZYSKIE','ŚLĄSKIE',
        'POMORSKIE','PODLASKIE','PODKARPACKIE','OPOLSKIE','MAZOWIECKIE','MAŁOPOLSKIE','ŁÓDZKIE',
        'LUBUSKIE','LUBELSKIE', 'KUJAWSKO-POMORSKIE','DOLNOŚLĄSKIE']
```

```python
roznica = []
for i in range(len(srednia)):
    roznica.append(zmarli_2020[i] - srednia[i])
print(roznica)

df = pd.DataFrame({"names" : names,
                    "roznica" : roznica})

df = df.sort_values('roznica')
```

```python
plt.barh("names", "roznica", data = df)
plt.title("Wzrost liczby osób zmarłych w październiku 2020 względem średniej z 2015-2019", fontsize=15)
plt.show()
```

Matplotlib pozwala na proste tworzenie wykresów w stylu xkcd, służy do tego funkcja `xkcd()` z `matplotlib.pyplot`.
Warto zajrzeć do dokumentacji: https://matplotlib.org/gallery/showcase/xkcd.html
Mój przykład umieściłam poniżej.

Uwaga, w szczególności styl działa też z wykresami kołowymi i słupkowymi.

### Przykład

```python
x = [0, 0.05, 0.06, 0.9, 1]
y = [0, 0.5, 0, 0, 4]

# rysowanie w xkcd mode
with plt.xkcd():
    # ustawienie rozmiaru
    plt.figure(figsize=(10,10))
    
    plt.plot(x,y)
    
    # usuniecie podpisow osi
    plt.xticks([])
    plt.yticks([])
    
    # dodanie napisow
    plt.title("TWD project")
    plt.xlabel('Time')
    plt.ylabel('Productivity')

    plt.annotate(
        'One day untli deadline',
        xy=(0.9, 0), arrowprops=dict(arrowstyle='->'), xytext=(0.5, 1))
plt.show()
```

```python
with plt.xkcd():
    
    names = ["Chęci", "Umiejętnosci"]
    values = [100, 0]
    plt.bar(names, values)
    plt.title("Tworzenie plakatu na TWD")
```
