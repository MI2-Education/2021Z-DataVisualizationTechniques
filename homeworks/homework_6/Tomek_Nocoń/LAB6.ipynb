```python
import matplotlib.pyplot as plt
import numpy as np
```

# zad 1

```python
x = np.linspace(-3, 3, num = 50)
graph1 = np.power(x,3) + np.ones(50)
```

```python
x = np.linspace(-3, 3, num = 50)
graph2 = 4*np.sin(x) - np.ones(50) 
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('Charts')
ax1.plot(x, graph1, color = "blue")
ax2.plot(x, graph2, color = "orange")
plt.show()
```

```python
plt.title("Charts", loc = "center")
plt.plot(x, graph1, label = "x^3 + 1", color = "blue")
plt.plot(x, graph2, label = "4sinx + 1", color = "orange", linewidth = 3)
plt.legend()
plt.show()
```

```python
wojewodztwo = ["ZACHODNIOPOMORSKIE", "WIELKOPOLSKIE", "WARMINSKO-MAZURSKIE", "SWIETOKRZYSKIE",  "SLASKIE",
               "POMORSKIE", "PODLASKIE", "PODKARPACKIE", "OPOLSKIE", "MAZOWIECKIE", "MALOPOLSKIE", "LODZKIE",
               "LUBUSKIE", "LUBELSKIE", "KUJAWSKO-POMORSKIE", "DOLNOSLASKIE"]

l_2020 = [1874, 4013, 1565, 2008, 6270, 2521, 1468, 2909, 1356, 6564, 4743, 3687, 1313, 2731, 2435, 3665]

l_2015_2019 = [1509, 2796, 1241, 1208, 4255, 1830, 1049, 1643, 881, 4811, 2690, 2578, 894, 1920, 1768, 2683]
```

```python
frac = np.divide((np.subtract(l_2020, l_2015_2019)), l_2015_2019)*100
frac = np.sort(frac)
plt.barh(wojewodztwo, frac, color = "green")
plt.title("Procentowy przyrost zgonów w 2020 w prównianu do średinej z lat 2015-2019", loc = "center")
plt.show()
```
