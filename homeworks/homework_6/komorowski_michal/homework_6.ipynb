### Zadanie 1

a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

a)

```python
import numpy as np

fig = plt.subplots()

x = np.linspace(-3, 3, 100)
y = x*x*x + 1

ax1 = plt.plot(x, y, 'b', linewidth = 2)

plt.axis([-4, 4, -26, 28])
plt.grid(True)
plt.title(r'Wykres $y = x^3 + 1$ na przedziale [-3 , 3]')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$', rotation=0)

plt.show()
```

```python
y1 = x*x*x + 1
y2 = 2*x*x - 10

ax1 = plt.plot(x, y1, 'b', linewidth = 1, label=r'$y = x^3 + 1$')
ax2 = plt.plot(x, y2, 'r', linewidth = 2, label=r'$y = 2x^2-10$')

plt.axis([-3, 3, -26, 28])
plt.grid(True)
plt.title(r'Wykres $y = x^3 + 1$ oraz $y = 2x^2-10$ na przedziale [-3 , 3]')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$', rotation=0)
plt.legend()

plt.show()
```

c)

```python
fig = plt.figure()


ax1 = fig.add_subplot(1, 2, 1)
plt.plot(x, y1, 'b', linewidth = 1, label=r'$y = x^3 + 1$')


plt.axis([-4, 4, -26, 28])
plt.grid(True)
plt.title(r'Wykres $y = x^3 + 1$' '\n' 'na przedziale [-3 , 3]')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$', rotation=0)

ax2 = fig.add_subplot(1, 2, 2)
plt.plot(x, y1, 'b', linewidth = 1, label=r'$y = x^3 + 1$')
plt.plot(x, y2, 'r', linewidth = 2, label=r'$y = 2x^2-10$')

plt.axis([-3, 3, -26, 28])
plt.grid(True)
plt.title(r'Wykres $y = x^3 + 1$' '\n' 'oraz $y = 2x^2-10$' '\n' 'na przedziale [-3 , 3]')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$', rotation=0)
plt.legend()

plt.tight_layout()
```

# Zadanie 3

Poprawić wykres tak, aby łatwiej było odczytać w którym województwie śmiertelność wzrosła najbardziej.
Jest przynajmniej kilka innych podejść: scatterplot z latami na osi OX/OY, odcinki zaczynające się w danych ze średniej a kończące w 2020.

![https://static.im-g.pl/im/7/26498/m26498997,WOJEWODZTWA2.png](https://static.im-g.pl/im/7/26498/m26498997,WOJEWODZTWA2.png)

Źródło: https://static.im-g.pl/im/7/26498/m26498997,WOJEWODZTWA2.png

```python
wojewodztwo = ["ZACHODNIOPOMORSKIE", "WIELKOPOLSKIE", "WARMINSKO-MAZURSKIE", "SWIETOKRZYSKIE",  "SLASKIE", "POMORSKIE", "PODLASKIE", "PODKARPACKIE", "OPOLSKIE", "MAZOWIECKIE", "MALOPOLSKIE", "LODZKIE", "LUBUSKIE", "LUBELSKIE", "KUJAWSKO-POMORSKIE", "DOLNOSLASKIE"]

l_2020 = [1874, 4013, 1565, 2008, 6270, 2521, 1468, 2909, 1356, 6564, 4743, 3687, 1313, 2731, 2435, 3665]
l_2015_2019 = [1509, 2796, 1241, 1208, 4255, 1830, 1049, 1643, 881, 4811, 2690, 2578, 894, 1920, 1768, 2683]

# Myślę, że do pokazania tego co chcemy lepsze będzie różnica liczby śmierci w odniesieniu do ludności np. na 100 tys. mieszkańcow, ponieważ np. w województwie mazowieckim zawsze umrze więcej osób niż w podlaskim w związku z duża różnicą w liczbie ludności, a zaburza to bardzo proporcje i czytelność na wykresie.

ludnosc = [1710482, 3475323, 1439675, 1257179, 4570849, 2307710, 1188800, 2127657, 996011, 5349114, 3372618, 2493603, 1018075, 2139726, 2086210, 2904207]

l_2015_2019_norm = [i / (j / 100000) for i, j in zip(l_2015_2019, ludnosc)]
l_2020_norm = [i / (j / 100000) for i, j in zip(l_2020, ludnosc)]

delta = [i - j for i, j in zip(l_2020_norm, l_2015_2019_norm)]

plt.barh(wojewodztwo, delta, zorder = 2.5)
plt.gca().xaxis.grid(True)
plt.xticks(np.arange(0, max(delta) + 5, 5))
plt.xlabel("Przyrost liczby zmarłych osób [na 100 tys. mieszkańców]")

plt.title("Różnica liczby osób zmarłych na 100 tys. mieszkańców w poszczególnych \n województwach w październiku 2020 r. i średnio w latach 2015-2019")

plt.show()
```

Matplotlib pozwala na proste tworzenie wykresów w stylu xkcd, służy do tego funkcja `xkcd()` z `matplotlib.pyplot`.
Warto zajrzeć do dokumentacji: https://matplotlib.org/gallery/showcase/xkcd.html
Mój przykład umieściłam poniżej.

Uwaga, w szczególności styl działa też z wykresami kołowymi i słupkowymi.

### Przykład

```python
x = [0, 0.05, 0.06, 0.9, 1]
y = [0, 0.5, 0, 0, 4]

# rysowanie w xkcd mode
with plt.xkcd():
    # ustawienie rozmiaru
    plt.figure(figsize=(10,10))
    
    plt.plot(x,y)
    
    # usuniecie podpisow osi
    plt.xticks([])
    plt.yticks([])
    
    # dodanie napisow
    plt.title("TWD project")
    plt.xlabel('Time')
    plt.ylabel('Productivity')

    plt.annotate(
        'One day untli deadline',
        xy=(0.9, 0), arrowprops=dict(arrowstyle='->'), xytext=(0.5, 1))
plt.show()
```

```python
with plt.xkcd():
    x = ["Actual plot", "Changing colors and theme"]
    y = [20, 80]

    plt.figure(figsize=(7, 7))
    plt.pie(y, labels = x, autopct = '%1.0f%%', startangle = 90, colors = ["black", "white"], wedgeprops = {"edgecolor": "black", 'linewidth': 1})
    plt.title("Time spent on making a plot")

    plt.show()
```
