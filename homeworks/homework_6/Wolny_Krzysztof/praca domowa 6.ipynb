```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Zadanie 1


a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

```python
x = np.linspace(-3, 3, 50)
y = x**3 + 1
```

```python
plt.plot(x, y)
plt.ylabel('y = x^3 + 1')
plt.xlabel('x')
plt.title("y = x^3 + 1", loc = "center", size = 18)
plt.show()
```

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

```python
y2 = x**2 + 1
```

```python
plt.plot(x, y, label = "x^3 + 1")
plt.plot(x, y2, color = "red", linewidth=3, label = "x^2 + 1")
plt.ylabel('y')
plt.xlabel('x')
plt.legend()
plt.title("Comparison x^3 + 1 and x^2 +1", loc = "center", size = 18)
plt.show()
```

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

```python
fig = plt.figure()
ax = fig.add_subplot(2, 1, 1)
fig.suptitle('Comparison x^3 + 1 and x^2 +1')
ax.plot(x, y)
ay = fig.add_subplot(2, 1, 2)
ay.plot(x, y2, color = "red", linewidth=3, label = "x^2 + 1")
plt.show()
```

## Zadanie2

a) Narysować wykres słupkowy na podanych danych.

```python
labels = ['Blue', 'Red', 'Green', 'Purple']
sizes = [15, 30, 45, 10]
```

```python
plt.bar(labels, sizes, color = ["turquoise", "red", "green", "purple"])
plt.title('Colors', size = 16)
plt.show()
```

b) Znaleźć sposób (w internecie) na narysowanie wykresu kołowego na podanych danych.

```python
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90, 
        colors = ["turquoise", "red", "green", "purple"])
ax1.axis('equal')  
fig1.suptitle('Colors', size = 16)
plt.show()
```

## Zadanie 3

Poprawić wykres tak, aby łatwiej było odczytać w którym województwie śmiertelność wzrosła najbardziej. Jest przynajmniej kilka innych podejść: scatterplot z latami na osi OX/OY, odcinki zaczynające się w danych ze średniej a kończące w 2020. Jestem też ciekawa Waszych innych pomysłów.

![](./chart.png)

```python
wojewodztwo = ["ZACHODNIOPOMORSKIE", "WIELKOPOLSKIE", "WARMINSKO-MAZURSKIE", "SWIETOKRZYSKIE",  "SLASKIE", "POMORSKIE", "PODLASKIE", "PODKARPACKIE", "OPOLSKIE", "MAZOWIECKIE", "MALOPOLSKIE", "LODZKIE", "LUBUSKIE", "LUBELSKIE", "KUJAWSKO-POMORSKIE", "DOLNOSLASKIE"]

l_2020 = [1874, 4013, 1565, 2008, 6270, 2521, 1468, 2909, 1356, 6564, 4743, 3687, 1313, 2731, 2435, 3665]

l_2015_2019 = [1509, 2796, 1241, 1208, 4255, 1830, 1049, 1643, 881, 4811, 2690, 2578, 894, 1920, 1768, 2683]
```

Można to zrobic tak jak w poleceniu, czyli zrobić scatterplot z odcinkami łączacymi wartość średniej z danymi z 2020. 

```python
plt.title('Porównanie liczby zmałych w poszczególnych województwch \n w paździeniuku 2020r. i średnio w latach 2015-2019', 
         loc = 'left')
plt.scatter(l_2020, wojewodztwo, color = "red")
plt.scatter(l_2015_2019, wojewodztwo, color = "black")
for i in range(0, len(wojewodztwo)):
    x = [l_2015_2019[i], l_2020[i]]
    plt.plot(x, [wojewodztwo[i], wojewodztwo[i]], color = "black")                
plt.show()
```

Możemy równiez pokazać wzrost procentowy między zmarłymi w latach 2015-2019 i w 2020 roku na wykresie słupkowym. 

```python
import matplotlib.ticker as mtick
sub = list((np.array(l_2020) - np.array(l_2015_2019))/np.array(l_2015_2019)*100)
df = pd.DataFrame({"Wojewodztwo":wojewodztwo,
                  "Roznica":sub})
df = df.sort_values("Roznica")

fig = plt.figure(figsize=(9,5))
fig.suptitle('Procentowy wzrost zmałych w poszczególnych województwch \n w paździeniuku 2020r. i średnio w latach 2015-2019', 
            ma = 'left', 
            fontsize = 'large')

ax = plt.subplot(1, 1, 1)
ax.grid(color='grey', linestyle='-', 
         linewidth=1, 
         axis = 'x', 
        )
ax.set_axisbelow(True)
ax.xaxis.set_major_formatter(mtick.PercentFormatter())

ax.barh("Wojewodztwo", "Roznica", data=df)
plt.xlim(0, 80)

plt.show()
```
