### Zadanie 1

a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

```python
x = np.linspace(-3, 3, 200)
y = np.power(x, 3) + 1
plt.plot(x, y, label = "x^3 + 1")
plt.suptitle("Wykres")
plt.xlabel("x")
plt.ylabel("y")
y_2 = 20 * np.sin(4 * x)
plt.plot(x, y_2, color = "red", linestyle = "--", linewidth = 3, label = "20sin(4x)")
plt.legend()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('Subplot')
ax1.plot(x, y, label = "x^3 + 1")
ax2.plot(x, y_2, color = "red", linestyle = "--", linewidth = 3, label = "20sin(4x)")
ax1.legend()
ax2.legend()
ax1.set_ylabel("y")
ax2.set_ylabel("y")
plt.show()
```

# Zadanie 3

Poprawić wykres tak, aby łatwiej było odczytać w którym województwie śmiertelność wzrosła najbardziej.
Jest przynajmniej kilka innych podejść: scatterplot z latami na osi OX/OY, odcinki zaczynające się w danych ze średniej a kończące w 2020.
Jestem też ciekawa Waszych innych pomysłów.

![https://static.im-g.pl/im/7/26498/m26498997,WOJEWODZTWA2.png](https://static.im-g.pl/im/7/26498/m26498997,WOJEWODZTWA2.png)

Źródło: https://static.im-g.pl/im/7/26498/m26498997,WOJEWODZTWA2.png

```python
wojewodztwo = ["ZACHODNIOPOMORSKIE", "WIELKOPOLSKIE", "WARMINSKO-MAZURSKIE", "SWIETOKRZYSKIE",  
               "SLASKIE", "POMORSKIE", "PODLASKIE", "PODKARPACKIE", 
               "OPOLSKIE", "MAZOWIECKIE", "MALOPOLSKIE", "LODZKIE", 
               "LUBUSKIE", "LUBELSKIE", "KUJAWSKO-POMORSKIE", "DOLNOSLASKIE"]

l_2020 = [1874, 4013, 1565, 2008, 6270, 2521, 1468, 2909, 
          1356, 6564, 4743, 3687, 1313, 2731, 2435, 3665]

l_2015_2019 = [1509, 2796, 1241, 1208, 4255, 1830, 1049, 1643, 
               881, 4811, 2690, 2578, 894, 1920, 1768, 2683]

# W celu porównywania zmian liczby osób zmarłych wyznaczymy procentowy przyrost

prz = [None] * len(l_2020)
for i in range(len(prz)):
    prz[i] = (l_2020[i] - l_2015_2019[i]) * 100 / l_2015_2019[i]

plt.figure(figsize=(10, 10))
plt.barh(wojewodztwo, prz, color = "red")
plt.xlim(0, 100)
for i in range(20, 81, 20):
    plt.axvline(i, color='black', linestyle='--', linewidth=0.3)
plt.title("Przyrost liczby osób zmarłych w październiku 2020r.\n\
    w porównaniu do średniej liczby osób zmarłych w październiku w latach 2015-2019")
plt.ylabel("WOJEWÓDZTWO")
plt.xlabel("WARTOŚĆ PROCENTOWA")
plt.show()
```

Matplotlib pozwala na proste tworzenie wykresów w stylu xkcd, służy do tego funkcja `xkcd()` z `matplotlib.pyplot`.
Warto zajrzeć do dokumentacji: https://matplotlib.org/gallery/showcase/xkcd.html
Mój przykład umieściłam poniżej.

Uwaga, w szczególności styl działa też z wykresami kołowymi i słupkowymi.

```python
labels = ["PlayStation 1", "PlayStation 2", "PlayStation 3", "PlayStation 4", "PlayStation 5"]
values = [1, 2, 3, 4, 5]
with plt.xkcd():
    plt.figure(figsize=(10,10))
    plt.bar(labels, values)
    plt.title("PLAYSTATION")
```
