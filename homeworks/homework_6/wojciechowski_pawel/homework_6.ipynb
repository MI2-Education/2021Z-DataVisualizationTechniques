### Zadanie 1

a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

```python
# a)
import numpy as np

x = np.linspace(-3, 3, 100)
y = x**3 + 1

plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = x^3+1", loc = "center")
plt.show()
```

```python
# b)
y2 = -(x**2)+6
plt.plot(x,y, label = "x^3+1")
plt.plot(x,y2, label = "-(x^2)+6", linewidth = 4)
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Dwie funkcje", loc = "center")
plt.show()
```

```python
# c)
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x,y)
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.title.set_text("y = x^3+1")

ax2.plot(x,y, label = "x^3+1")
ax2.plot(x,y2, label = "-(x^2)+6", linewidth = 4)
ax2.legend()
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.title.set_text("Dwie funkcje")

plt.show()
```

Jestem też ciekawa Waszych innych pomysłów.

![https://static.im-g.pl/im/7/26498/m26498997,WOJEWODZTWA2.png](https://static.im-g.pl/im/7/26498/m26498997,WOJEWODZTWA2.png)

Źródło: https://static.im-g.pl/im/7/26498/m26498997,WOJEWODZTWA2.png

```python
wojewodztwo = ["ZACHODNIOPOMORSKIE", "WIELKOPOLSKIE", "WARMINSKO-MAZURSKIE", "SWIETOKRZYSKIE",  "SLASKIE", "POMORSKIE", "PODLASKIE", "PODKARPACKIE", "OPOLSKIE", "MAZOWIECKIE", "MALOPOLSKIE", "LODZKIE", "LUBUSKIE", "LUBELSKIE", "KUJAWSKO-POMORSKIE", "DOLNOSLASKIE"]

l_2020 = [1874, 4013, 1565, 2008, 6270, 2521, 1468, 2909, 1356, 6564, 4743, 3687, 1313, 2731, 2435, 3665]

l_2015_2019 = [1509, 2796, 1241, 1208, 4255, 1830, 1049, 1643, 881, 4811, 2690, 2578, 894, 1920, 1768, 2683]

roznica = [x - y for x,y in zip(l_2020, l_2015_2019)]
procentowa_roznica = [round((x/y-1)*100, 1) for x,y in zip(l_2020, l_2015_2019)]
```

```python
import pandas as pd

df = pd.DataFrame({'wojewodztwo': wojewodztwo, 'l_2020':l_2020, 'l_2015_2019':l_2015_2019, 'procent':procentowa_roznica})
df = df.sort_values('l_2020', ascending=False)

width_ = 0.3
x = np.arange(len(df.wojewodztwo))


plt.rcParams['figure.figsize'] = [14, 8]
#plt.bar(x-width/2,l_2020, color='#974868')
#plt.bar(x-width/2,l_2015_2019, color='#759290')
#plt.bar(x+width/2, procentowa_roznica)
#plt.xticks(rotation = 70)

plt.rcParams['figure.figsize'] = [14, 8]

fig, ax1 = plt.subplots()

plt.xticks(x, df.wojewodztwo, rotation = 60)
ax1.bar(x-width_/2,df.l_2020, width_, color='#f07564', label = 'Liczba zgonów w październiku 2020',edgecolor='black')
ax1.bar(x-width_/2,df.l_2015_2019, width_,color='#9c9c9c', label = "Średnia liczba zgonów w październikach 2015-2019",edgecolor='black')
ax1.set_ylabel('Liczba zgonóww', color = "#823426", fontsize = 14)
ax1.tick_params(axis = 'y', labelcolor = '#823426')
ax1.legend(loc=2)
ax1.set_ylim((0,7500))



ax2 = ax1.twinx()


ax2.bar(x+width_/2, df.procent, width_, color = "#6692f4", edgecolor='black', label = "Procentowa różnica pomiędzy \npaździernikiem 2020 a średnią lat 2015-2020")
ax2.set_ylabel("Różnica w procentach [%]", color = "#383c99", fontsize = 14)
ax2.tick_params(axis='y', labelcolor="#383c99")
ax2.legend(loc=0)
ax2.set_ylim((0,100))

plt.title("Porównianie liczby osób zmarłych w poszczególnych województwach, \n w październiku 2020 i średnio w latach 2015-2020", fontsize = 18)
plt.legend()
plt.show()

# ##ca5e4a
# #777acd
```

Matplotlib pozwala na proste tworzenie wykresów w stylu xkcd, służy do tego funkcja `xkcd()` z `matplotlib.pyplot`.
Warto zajrzeć do dokumentacji: https://matplotlib.org/gallery/showcase/xkcd.html
Mój przykład umieściłam poniżej.

Uwaga, w szczególności styl działa też z wykresami kołowymi i słupkowymi.

### Przykład

```python
x = [0, 0.05, 0.06, 0.9, 1]
y = [0, 0.5, 0, 0, 4]

# rysowanie w xkcd mode
with plt.xkcd():
    # ustawienie rozmiaru
    plt.figure(figsize=(10,10))
    
    plt.plot(x,y)
    
    # usuniecie podpisow osi
    plt.xticks([])
    plt.yticks([])
    
    # dodanie napisow
    plt.title("TWD project")
    plt.xlabel('Time')
    plt.ylabel('Productivity')

    plt.annotate(
        'One day untli deadline',
        xy=(0.9, 0), arrowprops=dict(arrowstyle='->'), xytext=(0.5, 1))
plt.show()
```

```python
x = [1,2,3,4]
with plt.xkcd():
    plt.figure(figsize=(10,10))
    plt.pie(x, colors=["#377eb8","#e41a1c","#4daf4a","#984ea3"], labels=['niebieski', 'czerwony', 'zielony','fioletowy'])
    plt.title("Procentowy udział kolorów na wykresie kołowym")
```
