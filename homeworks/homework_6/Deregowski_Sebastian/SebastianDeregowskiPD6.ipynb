# Zadania

### Zadanie 1

a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

```python
## podpunkt a
x = np.arange(-3,3,0.1)
y = x**3+1
plt.plot(x,y)
plt.title("Wykres y=x^3+1")
plt.ylabel("oś y")
plt.xlabel("oś x")
plt.show()
```

```python
## podpunkt b
x = np.arange(-3,3,0.1)
z = x**5-8*x**3-4
plt.plot(x,y,label="y=x^3+1")
plt.plot(x,z,label="y=x^5-8x^3-4",linewidth=3,color="orange")
plt.title("Wykresy x^3+1 i x^5-8x^3-4")
plt.ylabel("oś y")
plt.xlabel("oś x")
plt.legend()
plt.show()
```

```python
## podpunkt c
fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('Dwa wykresy oddzielnie')

ax1.plot(x, y,label="y=x^3+1")
ax2.plot(x, z,label="y=x^5-8x^3-4")
ax1.legend()
ax2.legend()
plt.show()
```

Matplotlib pozwala na proste tworzenie wykresów w stylu xkcd, służy do tego funkcja `xkcd()` z `matplotlib.pyplot`.
Warto zajrzeć do dokumentacji: https://matplotlib.org/gallery/showcase/xkcd.html
Mój przykład umieściłam poniżej.

Uwaga, w szczególności styl działa też z wykresami kołowymi i słupkowymi.

```python
x = [0, 0.1, 0.25, 0.3, 0.35, 0.5, 0.7, 0.75, 0.8, 0.95, 1]
y = [0, 0.1, 0,    1.5, 0,    0,   0,   1.2,  0,   0, 4]

# rysowanie w xkcd mode
with plt.xkcd():
    # ustawienie rozmiaru
    plt.figure(figsize=(10,10))
    
    plt.plot(x,y)
    
    # usuniecie podpisow osi
    plt.xticks([])
    plt.yticks([])
    
    # dodanie napisow
    plt.title("Stress level throughout the semester")
    plt.xlabel('Time')
    plt.ylabel('Stress level')
    
    plt.annotate(
        'Weekend before exam',
        xy=(1, 4), arrowprops=dict(arrowstyle='->'), xytext=(0.7, 3))
    
    plt.annotate(
        'Project deadline',
        xy=(0.75, 1.2), arrowprops=dict(arrowstyle='->'), xytext=(0.5, 1.5))
    
    plt.annotate(
        'Project announcement',
        xy=(0.5, 0), arrowprops=dict(arrowstyle='->'), xytext=(0.4, 0.8))
    
    plt.annotate(
        'Night before test',
        xy=(0.3, 1.5), arrowprops=dict(arrowstyle='->'), xytext=(0.3, 2))
    
    plt.annotate(
        'Test announcement',
        xy=(0.1, 0.1), arrowprops=dict(arrowstyle='->'), xytext=(0.01, 0.5))
    
plt.show()
```
