# Raport z eksploracji danych

#### Autorzy: Kinga Ułasik, Szymon Rećko, Hubert Ruczyński

## Pomysł na plakat

Podczas przeglądania podrzuconego pliku na temat pandemii koronawirusa w Polsce wpaliśmy na pomysł stworzenia plakatu, którego celem jest ukazanie jak poszczególne województwa radzą sobie z pandemią.Chcemy wyróżnić po trzy województwa, które radzą sobie najlepiej oraz najgorzej z tą trudną sytuacją.

## Eksploracja danych

Oczywiście, aby być w stanie wyłonić najlepsze i najgorsze województwa musimy wyróżnić cechy które uważamy za najistotniejsze, aby dokonać ich ewaluacji. Te informacje uzyskaliśmy podczas eksploracji danych zawartych w bazie danych: Covid w Polsce.

Aby dane były kompletne postanowiliśmy zbadać stan województw na dzień 16.11, gdyż jest to ostatnia data, dla której uzyskaliśmy kompletne dane. Na początku wyróżniliśmy następne dane dotyczące rozwoju pandemii w województwach jako bardzo przydatne dla naszego badania: Tygodniowa liczba przebadanych próbek w woj., Suma przebadanych próbek w woj., Liczba testów na 1000 mieszkańców, Populacja na 2019 GUS,  Suma osób z wynikiem pozytywnym, Suma wyzdrowień, Suma zgonów, Liczba osób hospitalizowanych, Liczba łóżek dla pacjentów, % zajęte łóżka, Liczba zajętych respiratorów (stan ciężki), Liczba respiratorów, % zajęte repiratorów, Nowe wyzdrowienia wegług województwa, Przyrost dzienny pozytywnych przypadków, Średnia nowych zgonów z tygodnia

Po głębszym zastanowieniu się nad nimi, doszliśmy do wniosku, że pięcioma najważniejszymy dla nas danymi będą: liczba łóżek i repiratorów, oraz % ich zajęcia, liczba testów na 1000 mieszkańców, suma osób z wynikiem pozytywnym, suma wyzdrowień oraz suma zgonów.

Z tych danych wyróżniliśmy 5 kryteriów, które w rónym stopniu wpływają na ocenę danego województwa. Są to Testowanie, Przygotowanie Szpitali, Rzeczywiste wykorzystanie szpitali, Procentowa liczba chorych oraz Skutek choroby. Aby wyznaczyć najlepsze i najgorsze województwa przygotowaliśmy też punktację ze względu na każde kryterium. Punkty wyliczane są względem najlepszego i najgorszego województwa, tak aby skala punktacji rozpoczynałą się w 0, a kończyła w 100 dla każdej ze statystyk (sumarycznie województwo może dostać od 0 do 500 punktów)

## Wstępne wizualizacje

Testowanie

W jasny sposób widać, które województwa przeprowadzają najwięcej testów (ciemne kolory), a które najmniej (jasne kolory)

Przygotowanie Szpitali

Na powyższym wykresie, im bliżej lewego dolnego rogu wykresu znajduje się województwo, tym lepsze jest jego przygotowanie szpitali.

Rzeczywiste przygotowanie szpitali

Na powyższym wykresie, im bliżej lewego dolnego rogu wykresu znajduje się województwo, tym lepsza jest aktualna sytuacja w szpitalach.

Procentowa liczba chorych

Na powyższym wykresie przedstawiona jest procentowa liczba osób chorych w województwie.

Skutek choroby

Na powyższym wykresie, im bliżej prawego dolnego rogu wykresu znajduje się województwo, tym lepsze jest zakończenie choroby (czy osoba wyzdrowiała czy zmarła).

Najlepsze województwa

Na powyższym radar-charcie porównane są trzy najlepsze województwa (te które zdobyły  sumarycznie najwięcej punktów). Co ciekawe widać, że każde z nich posiada różne mocne i słabe strony.

Najgorsze województwa

Na powyższym radar-charcie porównane są trzy najgorsze województwa (te które zdobyły  sumarycznie najmniej punktów). Co ciekawe widać, że każde z nich także posiada różne mocne i słabe strony.
